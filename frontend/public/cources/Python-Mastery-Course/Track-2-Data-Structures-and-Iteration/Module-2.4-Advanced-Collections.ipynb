## Part 1 · `namedtuple` — Readable Records

A plain tuple stores values by position, so `point[0]` and `point[1]` give no hint of meaning. A `namedtuple` is a tuple subclass with **named fields**, so you can write `point.x` and `point.y`. It keeps a tuple's efficiency and immutability while reading far more clearly.

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y"])    # define a type with named fields
p = Point(3, 4)

print("by name:", p.x, p.y)        # readable access
print("by index still works:", p[0], p[1])
print("unpacking works:", (lambda a, b: a + b)(*p))

# It is still an immutable tuple. To get a changed copy, use _replace.
moved = p._replace(x=10)
print("original:", p, "| moved:", moved)
print("as a dict:", p._asdict())
```

## Part 2 · `deque` — Fast Operations at Both Ends

A list is efficient at its end but slow at its front, because inserting or removing at the front shifts every other element. A `deque` (double-ended queue) is designed for fast appends and pops at **both** ends, making it ideal for queues and sliding windows. It can also be given a maximum length, after which adding to one end discards from the other.

```python
from collections import deque

queue = deque(["a", "b", "c"])
queue.append("d")            # add at the right
queue.appendleft("z")        # add at the left (fast, unlike a list)
print("after additions:", queue)

print("pop right:", queue.pop())
print("pop left:", queue.popleft())
print("remaining:", queue)

# A bounded deque keeps only the most recent items, like a rolling history.
recent = deque(maxlen=3)
for n in range(1, 7):
    recent.append(n)        # older values fall off the left automatically
    print("window:", list(recent))
```

## Part 3 · `defaultdict` — Dictionaries with Automatic Defaults

A common pattern is to build a dictionary of lists or counts, which requires checking whether a key exists before using it. A `defaultdict` removes that step: you give it a factory (such as `list` or `int`), and any missing key is created with a fresh default the moment it is accessed.

```python
from collections import defaultdict

# Grouping with a plain dict requires setdefault or an existence check.
# With defaultdict(list), a missing key starts as an empty list automatically.
by_first_letter = defaultdict(list)
for word in ["apple", "avocado", "banana", "cherry", "blueberry"]:
    by_first_letter[word[0]].append(word)
print(dict(by_first_letter))

# defaultdict(int) makes counting concise: a missing key starts at 0.
tally = defaultdict(int)
for char in "mississippi":
    tally[char] += 1
print(dict(tally))
```

## Part 4 · `Counter` — Counting Made Trivial

`Counter` is a dictionary subclass purpose-built for counting. Pass it any iterable and it tallies occurrences. It offers `most_common`, supports arithmetic between counters, and treats missing keys as zero.

```python
from collections import Counter

votes = ["red", "blue", "red", "green", "blue", "red"]
counts = Counter(votes)
print("counts:", counts)
print("most common two:", counts.most_common(2))
print("count of a missing key:", counts["purple"])   # 0, no error

# Counters support arithmetic, useful for combining tallies.
monday = Counter(apples=3, pears=1)
tuesday = Counter(apples=1, pears=2, plums=5)
print("combined:", monday + tuesday)
print("difference:", tuesday - monday)
```

## Part 5 · `OrderedDict` and `ChainMap`

Since Python 3.7, regular dictionaries preserve insertion order, so `OrderedDict` is rarely needed for ordering alone. It remains useful when order-sensitive equality matters (two `OrderedDict`s are equal only if their order matches) and for its `move_to_end` method.

`ChainMap` groups several dictionaries into one view, searching them in order. It is handy for layered configuration, where user settings override defaults without copying anything.

```python
from collections import OrderedDict, ChainMap

od = OrderedDict([("a", 1), ("b", 2), ("c", 3)])
od.move_to_end("a")          # move key 'a' to the end
print("after move_to_end:", od)

# Two OrderedDicts with the same items but different order are NOT equal.
print("order-sensitive equality:",
      OrderedDict(a=1, b=2) == OrderedDict(b=2, a=1))   # False

# ChainMap: defaults underneath, overrides on top. Lookups search left to right.
defaults = {"theme": "light", "font": "sans", "size": 12}
user = {"theme": "dark"}
settings = ChainMap(user, defaults)
print("theme (overridden):", settings["theme"])   # from user
print("font (default):", settings["font"])         # falls through to defaults
```

### Example 1 — Records with namedtuple (Easy)

```python
from collections import namedtuple

Book = namedtuple("Book", ["title", "author", "year"])
b = Book("Dune", "Herbert", 1965)
print(f"{b.title} by {b.author} ({b.year})")
# Experiment: create another book and access its fields by name.
```

### Example 2 — A queue with deque (Easy)

```python
from collections import deque

line = deque()
line.append("Asha")          # people join at the back
line.append("Ben")
line.append("Chen")
print("now serving:", line.popleft())   # served from the front
print("waiting:", list(line))
```

### Example 3 — Grouping with defaultdict (Medium)

```python
from collections import defaultdict

orders = [("Asha", "book"), ("Ben", "pen"), ("Asha", "bag"), ("Chen", "pen")]
by_customer = defaultdict(list)
for customer, item in orders:
    by_customer[customer].append(item)

for customer, items in by_customer.items():
    print(f"{customer}: {items}")
```

### Example 4 — Counting with Counter (Medium)

```python
from collections import Counter

text = "the quick brown fox jumps over the lazy dog the end"
word_counts = Counter(text.split())
print("three most common:", word_counts.most_common(3))

# Count individual characters, ignoring spaces.
letters = Counter(c for c in text if c != " ")
print("most common letter:", letters.most_common(1))
```

### Example 5 — A sliding-window maximum with a bounded deque (Difficult)

```python
from collections import deque

def windowed_averages(values, size):
    """Yield the average of each consecutive window of the given size."""
    window = deque(maxlen=size)
    for value in values:
        window.append(value)
        if len(window) == size:          # only report once the window is full
            yield sum(window) / size

data = [10, 20, 30, 40, 50, 60]
print("3-window averages:", list(windowed_averages(data, 3)))
```

### Example 6 — Layered configuration with ChainMap (Difficult)

```python
from collections import ChainMap

site_defaults = {"lang": "en", "theme": "light", "items_per_page": 20}
team_policy = {"theme": "dark"}
user_prefs = {"items_per_page": 50}

# Order matters: user overrides team, which overrides defaults.
config = ChainMap(user_prefs, team_policy, site_defaults)
for key in ["lang", "theme", "items_per_page"]:
    print(f"{key} = {config[key]}")

# Writes go to the first mapping only, leaving the lower layers untouched.
config["lang"] = "fr"
print("user layer now:", user_prefs)
```

## Summary and Key Points

- `namedtuple` provides immutable records with named fields, combining tuple efficiency with readable access; `_replace` and `_asdict` help when needed.
- `deque` offers fast appends and pops at both ends and supports a maximum length for rolling windows, unlike a list which is slow at the front.
- `defaultdict` supplies a default for missing keys via a factory (`list`, `int`, and so on), simplifying grouping and counting.
- `Counter` tallies an iterable, supports `most_common` and arithmetic, and returns zero for absent keys.
- Regular dicts preserve insertion order, so `OrderedDict` is now used mainly for order-sensitive equality and `move_to_end`; `ChainMap` layers several mappings into one searchable view for configuration.
