## Part 1 · `sorted` and `list.sort`

There are two ways to sort. `sorted(iterable)` returns a **new** sorted list and works on any iterable. `list.sort()` sorts a list **in place** and returns `None`. Both accept `reverse=True` for descending order.

Python's sort is **stable**: items that compare equal keep their original relative order. This property makes multi-level sorting straightforward, as you will see shortly.

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print("sorted (new list):", sorted(numbers))
print("original unchanged:", numbers)
print("descending:", sorted(numbers, reverse=True))

numbers.sort()                 # in place, returns None
print("after list.sort():", numbers)

words = ["banana", "Apple", "cherry"]
print("default (case-sensitive):", sorted(words))         # uppercase sorts first
print("case-insensitive:", sorted(words, key=str.lower))  # a key fixes this
```

## Part 2 · Sorting with a `key` Function

The `key` argument names a function that is called on each element to produce the value used for comparison. The elements themselves are not changed; only the key decides order. This is how you sort by length, by a field, by absolute value, or by anything you can compute.

```python
words = ["pear", "fig", "banana", "kiwi"]
print("by length:", sorted(words, key=len))

numbers = [-5, 3, -1, 4, -2]
print("by absolute value:", sorted(numbers, key=abs))

# Sorting a list of tuples by a chosen element using a small lambda.
people = [("Asha", 30), ("Ben", 25), ("Chen", 35)]
print("by age:", sorted(people, key=lambda person: person[1]))
```

**Multi-level sorting** is achieved either with a key that returns a tuple (compared left to right) or, using stability, by sorting on the least significant field first and the most significant last.

```python
records = [
    ("Science", "Asha", 88),
    ("Maths", "Ben", 88),
    ("Science", "Chen", 91),
    ("Maths", "Dev", 75),
]

# Sort by subject ascending, then by score descending, using a tuple key.
# Negating the score reverses only that level.
ordered = sorted(records, key=lambda r: (r[0], -r[2]))
for row in ordered:
    print(row)
```

## Part 3 · `operator.itemgetter` and `attrgetter`

When a key simply pulls out an element or attribute, `operator.itemgetter` and `operator.attrgetter` express it more clearly and slightly faster than a lambda. `itemgetter` works with indices or dictionary keys; `attrgetter` works with object attributes. Both can extract several values at once for multi-level sorts.

```python
from operator import itemgetter, attrgetter
from collections import namedtuple

people = [("Asha", 30), ("Ben", 25), ("Chen", 35)]
print("by age via itemgetter:", sorted(people, key=itemgetter(1)))

dicts = [{"name": "Asha", "score": 88}, {"name": "Ben", "score": 92}]
print("by score via itemgetter:", sorted(dicts, key=itemgetter("score")))

Person = namedtuple("Person", ["name", "age"])
crowd = [Person("Asha", 30), Person("Ben", 25)]
print("by age via attrgetter:", sorted(crowd, key=attrgetter("age")))

# Multi-level with itemgetter: by score then name.
rows = [("Asha", 88), ("Ben", 88), ("Chen", 75)]
print("by score then name:", sorted(rows, key=itemgetter(1, 0)))
```

## Part 4 · `bisect` — Maintaining a Sorted List

If a list is already sorted, `bisect` finds where a value belongs using binary search, which is far faster than scanning. `bisect.bisect` returns the insertion index; `bisect.insort` inserts the value while keeping the list sorted. A classic use is converting a numeric score into a grade band.

```python
import bisect

scores = [10, 20, 30, 40, 50]

# Find where 35 would be inserted to keep the list sorted.
position = bisect.bisect(scores, 35)
print("35 would go at index:", position)

# Insert while preserving order, without re-sorting the whole list.
bisect.insort(scores, 35)
print("after insort:", scores)

# Map a value to a band using boundaries (a fast lookup table).
breakpoints = [60, 70, 80, 90]          # boundaries between grades
grades = "FDCBA"                          # five bands
for mark in [55, 65, 82, 91]:
    grade = grades[bisect.bisect(breakpoints, mark)]
    print(f"mark {mark} -> grade {grade}")
```

## Part 5 · `heapq` — Efficient Smallest and Largest

A heap keeps the smallest item instantly accessible while allowing efficient insertion and removal. The `heapq` module turns a plain list into a min-heap. It is the right tool for priority queues and for finding the top or bottom few items of a large dataset without fully sorting it, via `nlargest` and `nsmallest`.

```python
import heapq

data = [5, 1, 8, 3, 9, 2, 7]

heap = data.copy()
heapq.heapify(heap)              # rearrange into a heap in linear time
print("smallest available instantly:", heap[0])
print("popped in order:", [heapq.heappop(heap) for _ in range(3)])

# Top and bottom few without sorting everything.
print("three largest:", heapq.nlargest(3, data))
print("two smallest:", heapq.nsmallest(2, data))

# A priority queue: push (priority, task) tuples; lowest priority comes out first.
pq = []
heapq.heappush(pq, (2, "write report"))
heapq.heappush(pq, (1, "fix urgent bug"))
heapq.heappush(pq, (3, "reply to email"))
print("next task:", heapq.heappop(pq))
```

```python
students = [
    {"name": "Asha", "class": "B", "score": 88},
    {"name": "Ben", "class": "A", "score": 75},
    {"name": "Chen", "class": "A", "score": 88},
    {"name": "Dev", "class": "B", "score": 75},
]

# Sort by class ascending, then by score descending.
ordered = sorted(students, key=lambda s: (s["class"], -s["score"]))
for s in ordered:
    print(s["class"], s["name"], s["score"])
```

### Example 5 — Grade bands with bisect (Difficult)

```python
import bisect

def grade(mark, breakpoints=(50, 60, 70, 80, 90), bands="FEDCBA"):
    """Map a numeric mark to a letter band using binary search."""
    return bands[bisect.bisect(breakpoints, mark)]

for mark in [45, 58, 73, 88, 95]:
    print(f"{mark} -> {grade(mark)}")
# Experiment: change the breakpoints or bands and re-run.
```

### Example 6 — A task scheduler with heapq (Difficult)

```python
import heapq

class Scheduler:
    """A priority queue of tasks; lower numbers run first."""
    def __init__(self):
        self._heap = []
        self._counter = 0           # tiebreaker to keep insertion order for equal priorities

    def add(self, priority, task):
        heapq.heappush(self._heap, (priority, self._counter, task))
        self._counter += 1

    def run_next(self):
        priority, _, task = heapq.heappop(self._heap)
        return f"running '{task}' (priority {priority})"

s = Scheduler()
s.add(2, "backup")
s.add(1, "deploy")
s.add(2, "notify")     # same priority as backup; added later, runs later
s.add(0, "alert")

while s._heap:
    print(s.run_next())
```

```python
import bisect
timestamps = [5, 10, 15, 20, 25, 30]

def count_before(ts, t):
    return bisect.bisect_left(ts, t)   # index of t is the count of values below it

print(count_before(timestamps, 15))   # 2 (5 and 10)
print(count_before(timestamps, 31))   # 6
print(count_before(timestamps, 5))    # 0
```

## Summary and Key Points

- `sorted` returns a new list from any iterable; `list.sort` sorts in place and returns `None`. Both accept `reverse`, and the sort is stable.
- A `key` function decides order without altering elements; returning a tuple gives multi-level sorting, and stability allows sorting by successive fields.
- `operator.itemgetter` and `attrgetter` express element and attribute keys clearly and can extract several values for multi-level sorts.
- `bisect` uses binary search on a sorted list to find insertion points and insert in order, and is excellent for mapping values into bands.
- `heapq` turns a list into a min-heap for efficient priority queues and provides `nlargest` and `nsmallest` to find extremes without a full sort.
