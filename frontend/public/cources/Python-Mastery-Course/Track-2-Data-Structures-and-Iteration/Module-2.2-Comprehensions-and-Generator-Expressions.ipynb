## Part 1 · List Comprehensions

A list comprehension builds a list from an iterable in a single expression. The basic shape is:

```python
[expression for item in iterable]
```

It is read as: "the expression, for each item in the iterable." It is equivalent to a `for` loop that appends to a list, but shorter and often clearer.

````python
# A loop and the equivalent comprehension produce the same list.
squares_loop = []
for n in range(1, 6):
    squares_loop.append(n * n)

squares_comp = [n * n for n in range(1, 6)]

print("loop:", squares_loop)
print("comprehension:", squares_comp)
````

A condition can filter which items are included, by adding `if` at the end:

```python
[expression for item in iterable if condition]
```

A conditional **expression** can also appear at the front to choose the value itself, which is different from filtering.

````python
# Filtering: keep only even numbers.
evens = [n for n in range(10) if n % 2 == 0]
print("evens:", evens)

# Transforming with a choice: label each number, keeping all of them.
labels = ["even" if n % 2 == 0 else "odd" for n in range(6)]
print("labels:", labels)

# The two can combine: transform, then filter.
result = [n * 10 for n in range(10) if n % 3 == 0]
print("multiples of 3, times 10:", result)
````

## Part 2 · Dictionary and Set Comprehensions

The same idea builds dictionaries and sets. A dictionary comprehension uses `key: value` pairs inside braces; a set comprehension uses a single expression inside braces and removes duplicates.

````python
# Dictionary comprehension: map each word to its length.
words = ["sun", "moon", "stars"]
lengths = {word: len(word) for word in words}
print("lengths:", lengths)

# Build a lookup that inverts a mapping (value becomes key).
original = {"a": 1, "b": 2, "c": 3}
inverted = {value: key for key, value in original.items()}
print("inverted:", inverted)

# Set comprehension: distinct first letters, duplicates removed automatically.
names = ["Asha", "Ben", "Anil", "Bina", "Chen"]
initials = {name[0] for name in names}
print("initials:", initials)
````

## Part 3 · Nested Comprehensions

A comprehension can contain more than one `for`, which iterates over combinations, and can also nest a comprehension inside the expression to build a structure such as a grid. Read multiple `for` clauses left to right, exactly as you would write the equivalent nested loops.

````python
# Two for clauses: all combinations of two ranges (a flat list).
pairs = [(x, y) for x in range(1, 4) for y in range(1, 3)]
print("pairs:", pairs)

# A comprehension inside the expression builds a grid (a list of lists).
grid = [[row * col for col in range(1, 4)] for row in range(1, 4)]
for line in grid:
    print(line)

# Flattening a list of lists with two for clauses.
nested = [[1, 2], [3, 4], [5, 6]]
flat = [value for sublist in nested for value in sublist]
print("flattened:", flat)
````

> **Guidance.** Comprehensions are excellent for one clear transformation or filter. When logic grows (several conditions, side effects, or anything you must pause to decode), a plain loop is more readable. Clarity matters more than compactness.

## Part 4 · Generator Expressions

A generator expression looks like a list comprehension but uses parentheses. The crucial difference is that it does not build the whole result at once; it produces values **one at a time, on demand**. This is far more memory-efficient for large or infinite sequences, and it integrates naturally with functions such as `sum`, `max`, and `any`.

````python
# A list comprehension builds and stores every value.
list_version = [n * n for n in range(5)]
print("list:", list_version, type(list_version))

# A generator expression produces values lazily; printing it shows the object, not values.
gen_version = (n * n for n in range(5))
print("generator object:", gen_version)

# Consume it by iterating or by passing it to a function.
print("as a list:", list(gen_version))
# Note: a generator is exhausted after one pass; iterating again yields nothing.
print("second pass:", list(gen_version))
````

````python
import sys

# Memory comparison for one million squares.
big_list = [n * n for n in range(1_000_000)]
big_gen = (n * n for n in range(1_000_000))

print("list size (bytes):", sys.getsizeof(big_list))
print("generator size (bytes):", sys.getsizeof(big_gen))
# The generator stays tiny regardless of how many values it will produce.

# When you only need an aggregate, pass a generator directly; no list is built.
total = sum(n * n for n in range(1_000_000))   # parentheses optional as the sole argument
print("sum of squares:", total)
````

## Part 5 · When Each Is Appropriate

- Use a **list comprehension** when you need the full collection: to index it, reuse it, iterate more than once, or know its length.
- Use a **generator expression** when you will consume the values once, especially for large data or when feeding an aggregating function, to avoid building a list you do not need.

### Example 1 — Transform a list (Easy)

````python
prices = [100, 250, 80]
with_tax = [round(p * 1.18, 2) for p in prices]    # add 18 percent tax
print(with_tax)
# Experiment: change the tax rate or the prices.
````

### Example 2 — Filter a list (Easy)

````python
numbers = [12, 7, 30, 5, 18, 3]
big = [n for n in numbers if n >= 10]
print("ten or more:", big)
# Experiment: change the threshold.
````

### Example 3 — Build a lookup dictionary (Medium)

````python
students = ["Asha", "Ben", "Chen"]
scores = [88, 92, 79]

# zip pairs them; the comprehension turns pairs into key: value entries.
score_of = {name: score for name, score in zip(students, scores)}
print(score_of)
print("Ben scored", score_of["Ben"])
````

### Example 4 — Distinct values with a set comprehension (Medium)

````python
sentence = "the quick brown fox the lazy dog the end"

# Distinct words longer than three letters, lowercased.
notable = {word.lower() for word in sentence.split() if len(word) > 3}
print("notable words:", notable)
# Experiment: change the length condition.
````

### Example 5 — A multiplication grid and its flattening (Difficult)

````python
# Build a 5x5 multiplication table as a list of rows.
table = [[row * col for col in range(1, 6)] for row in range(1, 6)]
for line in table:
    print(" ".join(f"{value:3}" for value in line))   # format each cell in 3 columns

# Flatten it and find the distinct products.
products = {value for row in table for value in row}
print("distinct products count:", len(products))
````

### Example 6 — Generators for efficient aggregation (Difficult)

````python
# Imagine processing a large stream; we never build a list.
def reading_stream(n):
    """Yield n simulated sensor readings."""
    for i in range(n):
        yield (i * 7) % 100        # some varying value

# Pass a generator expression straight into aggregating functions.
n = 100_000
average = sum(r for r in reading_stream(n)) / n
highest = max(r for r in reading_stream(n))
count_high = sum(1 for r in reading_stream(n) if r > 90)   # count via a generator

print(f"average={average:.2f} highest={highest} readings_over_90={count_high}")
````

## Summary and Key Points

- A list comprehension `[expr for item in iterable if cond]` builds a list in one readable expression; a leading conditional expression chooses the value, while a trailing `if` filters items.
- Dictionary comprehensions use `key: value` pairs and set comprehensions use a single expression in braces, removing duplicates.
- Multiple `for` clauses iterate over combinations and can flatten nested data; a comprehension inside the expression builds grids. Prefer a loop when logic becomes complex.
- A generator expression uses parentheses and yields values lazily, using almost no memory; it is consumed once and is ideal for large data and aggregation.
- Choose a list comprehension when you need the whole collection, and a generator expression when you consume values once.
