## Part 1 · Lists

A list is an **ordered, mutable** sequence. It can hold values of any type, including a mix, and it can grow and shrink. Lists are the workhorse container of Python.

Indexing starts at 0; negative indices count from the end. Slicing, written `list[start:stop:step]`, returns a new list.

```python
fruits = ["apple", "banana", "cherry", "date"]

print("length:", len(fruits))
print("first:", fruits[0], "| last:", fruits[-1])
print("slice [1:3]:", fruits[1:3])          # banana, cherry (stop excluded)
print("every second:", fruits[::2])         # apple, cherry
print("reversed copy:", fruits[::-1])

# Lists are mutable: we can change an element in place.
fruits[1] = "blueberry"
print("after change:", fruits)
```

The common list methods fall into a few groups. Adding: `append` (one item at the end), `insert` (at a position), `extend` (several items). Removing: `pop` (by index, returns it), `remove` (by value), `clear`. Reordering: `sort`, `reverse`. Note that `sort` and `reverse` change the list **in place** and return `None`.

```python
nums = [3, 1, 4, 1, 5]

nums.append(9)               # add one at the end
nums.insert(0, 0)            # insert at the front
nums.extend([2, 6])          # add several
print("after additions:", nums)

removed = nums.pop()         # remove and return the last item
nums.remove(1)               # remove the first value equal to 1
print("after removals:", nums, "| popped:", removed)

nums.sort()                  # sorts in place, returns None
print("sorted:", nums)
print("count of 1:", nums.count(1), "| index of 4:", nums.index(4))
```

> **Caution — assignment does not copy.** Writing `b = a` makes `b` another name for the same list, so changes through one are visible through the other. To copy, use `a.copy()` or `a[:]`.

```python
a = [1, 2, 3]
b = a              # same list, two names
b.append(99)
print("a is affected:", a)        # [1, 2, 3, 99]

c = a.copy()       # an independent copy
c.append(0)
print("a unaffected by c:", a)    # unchanged
```

## Part 2 · Tuples

A tuple is an **ordered, immutable** sequence. Once created it cannot be changed, which makes tuples suitable for fixed collections of related values, dictionary keys, and returning several values from a function.

Tuples are written with commas; the parentheses are often optional. A single-element tuple needs a trailing comma.

```python
point = (3, 4)               # parentheses are conventional but the comma defines it
also_tuple = 5, 6            # equally valid
single = (7,)                # a one-element tuple; the comma is required
not_a_tuple = (7)            # this is just the integer 7

print(point, also_tuple, single, type(not_a_tuple))

# Tuples are immutable: point[0] = 9 would raise an error.
print("first coordinate:", point[0])
```

**Packing and unpacking** are central to idiomatic Python. Several values pack into a tuple, and a tuple unpacks into several names. A starred name absorbs the remaining items.

```python
# Unpacking
x, y = point
print("x:", x, "y:", y)

# Starred unpacking captures the middle or the rest.
first, *middle, last = [10, 20, 30, 40, 50]
print("first:", first, "| middle:", middle, "| last:", last)

# Swapping values uses packing and unpacking in one line.
a, b = 1, 2
a, b = b, a
print("swapped:", a, b)
```

## Part 3 · Dictionaries

A dictionary maps **keys to values**. It is unordered by meaning (though it preserves insertion order), fast to look up by key, and mutable. Keys must be hashable (strings, numbers, tuples); values can be anything.

```python
person = {"name": "Asha", "age": 30, "city": "Pune"}

print("name:", person["name"])           # direct access; errors if the key is missing
print("get missing:", person.get("email", "not provided"))   # safe access with default

person["age"] = 31                        # update an existing key
person["email"] = "asha@example.com"      # add a new key
print(person)

print("keys:", list(person.keys()))
print("values:", list(person.values()))
for key, value in person.items():         # iterate over pairs
    print(f"  {key} -> {value}")
```

Useful methods include `setdefault` (get a key, inserting a default if absent) and `update` (merge another dictionary in). Since Python 3.9, the `|` operator merges two dictionaries into a new one, and `|=` updates in place.

```python
counts = {"a": 1}
counts.setdefault("b", 0)        # inserts b: 0 because b was absent
counts.setdefault("a", 99)       # leaves a unchanged because a exists
print("after setdefault:", counts)

counts.update({"a": 10, "c": 3}) # merge another dict in (overwrites a)
print("after update:", counts)

defaults = {"theme": "light", "size": "M"}
chosen = {"size": "L"}
print("merged with |:", defaults | chosen)   # new dict; chosen wins on conflicts

# Dictionary comprehension: build a dict from an expression.
squares = {n: n * n for n in range(1, 6)}
print("squares:", squares)
```

## Part 4 · Sets

A set is an **unordered collection of unique values**. It removes duplicates automatically and supports fast membership testing and mathematical set operations. A `frozenset` is an immutable set, usable as a dictionary key or set member.

```python
numbers = [1, 2, 2, 3, 3, 3, 4]
unique = set(numbers)            # duplicates are discarded
print("unique:", unique)
print("3 in set:", 3 in unique)  # very fast membership test

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print("union (|):", a | b)               # all elements
print("intersection (&):", a & b)        # common elements
print("difference (-):", a - b)          # in a but not b
print("symmetric difference (^):", a ^ b)# in one but not both
```

```python
# Sets are mutable; frozensets are not.
s = {1, 2}
s.add(3)
s.discard(99)        # discard does not error if the value is absent
print("set now:", s)

fs = frozenset([1, 2, 3])
# fs.add(4) would raise an error because a frozenset cannot be modified.
print("frozenset usable as a dict key:", {fs: "ok"})
```

## Part 5 · Choosing the Right Container

A short guide:

- **List** when order matters and contents may change, or when duplicates are allowed.
- **Tuple** for a fixed group of values, dictionary keys, or returning several values.
- **Dictionary** for lookups by a meaningful key, and for representing records.
- **Set** for uniqueness and fast membership tests, or for set arithmetic.

Membership testing (`in`) is fast for sets and dictionaries and slow for long lists, which often guides the choice.

### Example 1 — Managing a to-do list (Easy)

```python
todo = []
todo.append("write notes")
todo.append("review code")
todo.insert(0, "check email")    # highest priority at the front
print("tasks:", todo)

done = todo.pop(0)               # complete the first task
print("completed:", done, "| remaining:", todo)
# Experiment: add more tasks and remove one by value with todo.remove(...).
```

### Example 2 — Returning several values with a tuple (Easy)

```python
def stats(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

low, high, mean = stats([4, 8, 15, 16, 23])     # unpack the returned tuple
print(f"low={low} high={high} mean={mean:.1f}")
# Experiment: change the list and observe the three values.
```

### Example 3 — Counting word frequency with a dictionary (Medium)

```python
text = "the cat sat on the mat the cat slept"
counts = {}
for word in text.split():
    counts[word] = counts.get(word, 0) + 1     # default 0 when first seen
print(counts)

most_common = max(counts, key=counts.get)      # key with the highest count
print("most common word:", most_common, "->", counts[most_common])
```

### Example 4 — Finding common and unique elements with sets (Medium)

```python
monday = {"Asha", "Ben", "Chen", "Dev"}
tuesday = {"Ben", "Dev", "Eva", "Fay"}

print("present both days:", monday & tuesday)
print("present either day:", monday | tuesday)
print("only on Monday:", monday - tuesday)
print("present exactly one day:", monday ^ tuesday)
# Experiment: adjust the names and observe the results.
```

### Example 5 — Grouping records with a dictionary of lists (Difficult)

```python
people = [
    ("Asha", "Pune"), ("Ben", "Delhi"), ("Chen", "Pune"),
    ("Dev", "Delhi"), ("Eva", "Pune"),
]

by_city = {}
for name, city in people:
    by_city.setdefault(city, []).append(name)   # start a list the first time a city appears
print(by_city)

for city, residents in by_city.items():
    print(f"{city}: {', '.join(residents)}")
```

### Example 6 — Merging and transforming with comprehensions (Difficult)

```python
prices = {"apple": 40, "banana": 20, "cherry": 80}
discount = {"banana": 15}                        # banana on offer

# Apply the discount where present, leaving others unchanged, in one comprehension.
final = {item: discount.get(item, price) for item, price in prices.items()}
print("final prices:", final)

# Keep only items priced above 30, as a new dict.
premium = {item: price for item, price in final.items() if price > 30}
print("premium items:", premium)

# A set of all distinct price values.
print("distinct prices:", set(final.values()))
```

## Summary and Key Points

- Lists are ordered and mutable; master indexing, slicing, and the add/remove/reorder methods. Remember that `sort` and `reverse` act in place and that assignment does not copy.
- Tuples are ordered and immutable, ideal for fixed groups, dictionary keys, and multiple return values. Packing and unpacking (including starred) are everyday tools.
- Dictionaries map keys to values with fast lookup; use `get`, `setdefault`, `update`, the `|` merge operator, and comprehensions.
- Sets store unique values with fast membership testing and support union, intersection, difference, and symmetric difference; frozensets are their immutable form.
- Choose the container by need: order and change favour lists, fixed groups favour tuples, keyed lookup favours dicts, and uniqueness or membership favours sets.
