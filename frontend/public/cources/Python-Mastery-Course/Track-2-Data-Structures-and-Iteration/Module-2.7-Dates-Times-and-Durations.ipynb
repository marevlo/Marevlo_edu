## Part 1 · `date`, `time`, and `datetime`

Three core types cover most needs. A `date` is a calendar day (year, month, day). A `time` is a clock time within a day (hour, minute, second). A `datetime` combines both. Each provides attributes such as `.year` and methods such as `.weekday()`.

`datetime.now()` gives the current local date and time. For reproducibility, the examples mostly use fixed dates rather than the live clock.

```python
from datetime import date, time, datetime

d = date(2025, 5, 30)
print("date:", d, "| year:", d.year, "| month:", d.month, "| day:", d.day)
print("weekday (Mon=0):", d.weekday(), "| iso weekday (Mon=1):", d.isoweekday())

t = time(14, 30, 15)
print("time:", t, "| hour:", t.hour, "| minute:", t.minute)

dt = datetime(2025, 5, 30, 14, 30, 15)
print("datetime:", dt)
print("its date part:", dt.date(), "| its time part:", dt.time())

# The live clock (value will differ each run):
print("current year is", datetime.now().year)
```

## Part 2 · `timedelta` — Durations and Arithmetic

A `timedelta` represents a span of time. Subtracting two dates or datetimes yields a `timedelta`; adding a `timedelta` to a date shifts it. This is the correct way to answer questions such as "what date is 90 days from now" or "how many days until a deadline," because it handles month lengths and leap years for you.

```python
from datetime import date, datetime, timedelta

start = date(2025, 1, 1)
end = date(2025, 5, 30)

duration = end - start                 # subtracting dates gives a timedelta
print("days between:", duration.days)

# Shifting a date forward and backward.
deadline = start + timedelta(days=90)
print("90 days after Jan 1:", deadline)
print("one week before the deadline:", deadline - timedelta(weeks=1))

# timedelta with time components.
meeting = datetime(2025, 5, 30, 9, 0)
print("meeting plus 2.5 hours:", meeting + timedelta(hours=2, minutes=30))

# A timedelta can be inspected and scaled.
span = timedelta(days=2, hours=6)
print("total seconds:", span.total_seconds())
```

## Part 3 · Formatting with `strftime` and Parsing with `strptime`

To turn a datetime into human-readable text, use `strftime` (string format time) with directive codes. To turn text back into a datetime, use `strptime` (string parse time) with a matching format. Common directives:

- `%Y` four-digit year, `%m` month number, `%d` day, `%H` hour (24), `%M` minute, `%S` second.
- `%A` weekday name, `%B` month name, `%p` AM or PM.

The format you pass to `strptime` must match the structure of the input text exactly.

```python
from datetime import datetime

dt = datetime(2025, 5, 30, 14, 5, 9)

print(dt.strftime("%Y-%m-%d"))                 # 2025-05-30
print(dt.strftime("%d %B %Y"))                 # 30 May 2025
print(dt.strftime("%A, %d %b %Y at %I:%M %p")) # Friday, 30 May 2025 at 02:05 PM

# Parsing text into a datetime; the format string describes the input.
parsed = datetime.strptime("30/05/2025 14:05", "%d/%m/%Y %H:%M")
print("parsed:", parsed, type(parsed))

# A mismatched format raises ValueError, which you would handle as in Module 1.5.
```

## Part 4 · ISO 8601 — The Reliable Exchange Format

ISO 8601 (for example `2025-05-30T14:05:09`) is an unambiguous, sortable text format and the right choice for storing or transmitting dates. Use `isoformat()` to produce it and `fromisoformat()` to read it. Because ISO strings sort chronologically as plain text, they are convenient in files and databases.

```python
from datetime import datetime

dt = datetime(2025, 5, 30, 14, 5, 9)
iso = dt.isoformat()
print("iso string:", iso)

back = datetime.fromisoformat(iso)            # parse it back exactly
print("round trip equal:", back == dt)

# ISO strings sort chronologically as text, which plain locale formats do not.
stamps = ["2025-05-30T09:00", "2025-05-30T08:30", "2025-01-02T23:59"]
print("sorted as text:", sorted(stamps))
```

## Part 5 · Time Zones with `zoneinfo`

A datetime without zone information is called **naive**; one with a zone is **aware**. Mixing them causes errors. The reliable discipline is: store and compute in **UTC**, and convert to a local zone only for display. The `zoneinfo` module (Python 3.9+) provides real time zones by name, including daylight-saving rules.

```python
from datetime import datetime, timezone
from zoneinfo import ZoneInfo

# An aware datetime in UTC. Prefer this for anything stored or compared.
now_utc = datetime(2025, 5, 30, 12, 0, tzinfo=timezone.utc)
print("UTC:", now_utc)

# Convert the same instant to other zones for display.
for zone in ["America/New_York", "Europe/London", "Asia/Kolkata", "Asia/Tokyo"]:
    local = now_utc.astimezone(ZoneInfo(zone))
    print(f"{zone:20} {local.strftime('%Y-%m-%d %H:%M %Z')}")

# The same instant, different wall-clock times; all represent one moment.
```

```python
from datetime import datetime
from zoneinfo import ZoneInfo

# Daylight saving is handled automatically: the UTC offset differs by season.
winter = datetime(2025, 1, 15, 12, 0, tzinfo=ZoneInfo("Europe/London"))
summer = datetime(2025, 7, 15, 12, 0, tzinfo=ZoneInfo("Europe/London"))
print("January offset:", winter.utcoffset())   # 0:00:00
print("July offset:", summer.utcoffset())       # 1:00:00 (British Summer Time)
```

### Example 2 — Days until a target (Easy)

### Example 3 — Formatting in several styles (Medium)

```python
from datetime import datetime

dt = datetime(2025, 5, 30, 18, 45)
styles = {
    "ISO date": "%Y-%m-%d",
    "Long": "%A, %d %B %Y",
    "12-hour clock": "%I:%M %p",
    "Compact": "%d%m%y",
}
for label, fmt in styles.items():
    print(f"{label:15} {dt.strftime(fmt)}")
```

### Example 4 — Parsing mixed input safely (Medium)

```python
from datetime import datetime

def parse_date(text):
    """Try a few known formats; return a datetime or None."""
    for fmt in ("%Y-%m-%d", "%d/%m/%Y", "%d %b %Y"):
        try:
            return datetime.strptime(text, fmt)
        except ValueError:
            continue
    return None

for sample in ["2025-05-30", "30/05/2025", "30 May 2025", "nonsense"]:
    print(f"{sample:14} -> {parse_date(sample)}")
```

### Example 5 — A simple age and countdown calculator (Difficult)

```python
from datetime import date

def years_between(born, on):
    """Whole years from 'born' to 'on', accounting for whether the birthday has passed."""
    years = on.year - born.year
    # Subtract one if the birthday has not yet occurred this year.
    if (on.month, on.day) < (born.month, born.day):
        years -= 1
    return years

born = date(1990, 7, 20)
today = date(2025, 5, 30)
print("age in whole years:", years_between(born, today))

# Days until the next birthday.
next_birthday = date(today.year, born.month, born.day)
if next_birthday < today:
    next_birthday = date(today.year + 1, born.month, born.day)
print("days until next birthday:", (next_birthday - today).days)
```

### Example 6 — Coordinating a meeting across time zones (Difficult)

```python
from datetime import datetime
from zoneinfo import ZoneInfo

# A meeting is set for 09:00 in New York. Show the local time elsewhere.
ny_time = datetime(2025, 5, 30, 9, 0, tzinfo=ZoneInfo("America/New_York"))
print("Scheduled (New York):", ny_time.strftime("%H:%M %Z"))

for city, zone in [("London", "Europe/London"),
                   ("Mumbai", "Asia/Kolkata"),
                   ("Tokyo", "Asia/Tokyo")]:
    local = ny_time.astimezone(ZoneInfo(zone))
    print(f"{city:8} {local.strftime('%H:%M %Z on %d %b')}")
```
