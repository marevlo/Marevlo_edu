## Part 1 · Conditional Statements

An `if` statement runs a block of code only when a condition is true. `elif` (else-if) adds further conditions, checked in order, and `else` provides a fallback when none matched. Indentation, not braces, defines which lines belong to a block; four spaces is the standard.

```python
score = 72

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"

print("Score", score, "earns grade", grade)
# Experiment: change score to 95 or 50 and re-run.
```

When a choice is simple, the **conditional expression** (often called the ternary operator) expresses it on one line: `value_if_true if condition else value_if_false`.

```python
temperature = 31
label = "warm" if temperature >= 25 else "cool"
print("It is", label)
```

## Part 2 · Loops

A `for` loop repeats once for each item in a sequence. A `while` loop repeats as long as a condition remains true.

`range(start, stop, step)` produces a sequence of integers up to but not including `stop`. `enumerate()` gives both an index and the item. `zip()` walks several sequences in parallel.

```python
# for over a range
total = 0
for i in range(1, 6):       # 1, 2, 3, 4, 5  (stop value 6 is excluded)
    total += i
print("Sum 1..5 =", total)

# while with a clear stopping condition
countdown = 3
while countdown > 0:
    print("tick", countdown)
    countdown -= 1
print("done")
```

```python
names = ["Asha", "Ben", "Chen"]

# enumerate gives index and value together (index starts at 0 by default).
for index, name in enumerate(names, start=1):
    print(index, name)

print("---")

# zip walks two lists in parallel, stopping at the shorter one.
scores = [88, 91, 79]
for name, score in zip(names, scores):
    print(name, "scored", score)
```

## Part 3 · `break`, `continue`, `pass`, and the loop `else`

`break` exits the nearest loop immediately. `continue` skips to the next iteration. `pass` does nothing and serves as a placeholder where a statement is syntactically required.

A loop may also have an `else` clause, which runs **only if the loop finished without hitting `break`**. It is ideal for search-style loops.

```python
# break and continue
for n in range(1, 11):
    if n % 2 == 0:
        continue            # skip even numbers
    if n > 7:
        break               # stop entirely once past 7
    print("kept:", n)
```

```python
# Loop else: detect whether a search succeeded.
numbers = [4, 9, 15, 22]
target = 7

for value in numbers:
    if value == target:
        print("Found", target)
        break
else:
    # Runs because the loop completed without breaking.
    print(target, "was not found")

# pass as a placeholder
for ch in "ab":
    pass   # nothing to do yet; this keeps the block syntactically valid
```

## Part 4 · Structural Pattern Matching (`match`/`case`, Python 3.10+)

The `match` statement compares a value against a series of **patterns** and runs the first that fits. It is far more powerful than a simple switch: patterns can match literals, sequences (capturing parts), mappings, and class shapes, and may carry a `guard` (an `if` condition). The wildcard `_` matches anything and acts as the default.

```python
def describe(command):
    match command:
        case "help":                       # literal pattern
            return "Showing help."
        case "quit" | "exit":              # OR pattern
            return "Goodbye."
        case ["move", x, y]:               # sequence pattern, capturing x and y
            return f"Moving to ({x}, {y})."
        case ["move", *rest]:              # sequence with a captured remainder
            return f"Move needs two values, got {len(rest)}."
        case {"action": action}:           # mapping pattern, capturing a value by key
            return f"Action requested: {action}."
        case n if isinstance(n, int) and n < 0:   # guard: a condition on the match
            return "Negative number."
        case _:                            # wildcard: the default case
            return "Unknown command."

for c in ["help", "quit", ["move", 3, 5], ["move", 1], {"action": "save"}, -8, 3.14]:
    print(c, "->", describe(c))
```

## Part 5 · The Walrus Operator (`:=`, Python 3.8+)

The walrus operator assigns a value to a name **as part of an expression**, so you can compute and test in one step. It is most useful in `while` loops and comprehensions, where it removes repetition.

```python
# A stream of readings; we process until the sentinel value 0 appears.
readings = iter([5, 8, 3, 0, 9])   # iter() lets us pull values one at a time with next()

# Without walrus we would call next() twice or restructure. With walrus it is one line:
while (value := next(readings)) != 0:
    print("processing", value)
print("stopped at sentinel")

# In a comprehension: keep the squared value only when it exceeds 10, without squaring twice.
results = [squared for x in range(6) if (squared := x * x) > 10]
print(results)
```

### Example 1 — Classifying a number (Easy)

Uses `if`/`elif`/`else`.

```python
n = 0
if n > 0:
    print("positive")
elif n < 0:
    print("negative")
else:
    print("zero")
# Experiment: change n to a positive or negative value.
```

### Example 2 — Summing with a loop, labelling with a ternary (Easy)

```python
numbers = [3, 8, 11, 6]
total = 0
for value in numbers:
    total += value
print("Total:", total, "(" + ("even" if total % 2 == 0 else "odd") + ")")
# Experiment: add or remove numbers and re-run.
```

### Example 3 — FizzBuzz (Medium)

A classic exercise combining loops, conditionals, and the remainder operator.

```python
for n in range(1, 21):
    if n % 15 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
# Note: check divisibility by 15 first, otherwise multiples of 15 would match 3 or 5 alone.
```

### Example 4 — A ranked leaderboard with `enumerate` and `zip` (Medium)

```python
players = ["Nadia", "Omar", "Priya"]
points = [120, 95, 140]

# Pair players with points, sort by points (highest first), then number them.
ranked = sorted(zip(players, points), key=lambda pair: pair[1], reverse=True)

for position, (player, score) in enumerate(ranked, start=1):
    print(position, player, score)
# Experiment: change the points and observe the new ordering.
```

### Example 5 — A small command parser with `match`/`case` (Difficult)

```python
def run(command):
    match command.split():                 # split the text into a list of words
        case ["go", direction]:
            return f"Walking {direction}."
        case ["take", *items] if items:     # guard ensures at least one item
            return f"Taking: {', '.join(items)}."
        case ["look"]:
            return "You see a quiet room."
        case _:
            return "I do not understand."

for text in ["go north", "take key map", "look", "dance"]:
    print(repr(text), "->", run(text))
```

### Example 6 — Reading until a sentinel with the walrus operator (Difficult)

```python
# Simulate a sequence of typed lines ending with an empty line.
lines = iter(["apple", "banana", "cherry", ""])

words = []
while (line := next(lines)) != "":      # assign and test in one expression
    words.append(line)

print("Collected", len(words), "words:", words)
# Experiment: add more fruit before the empty string.
```

**Solution 4**

```python
def status(code):
    match code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500 | 502:
            return "Server Error"
        case _:
            return "Unknown"

for c in [200, 404, 502, 301]:
    print(c, status(c))
```

**Solution 5**

```python
data = iter(["one", "", "two", "three", "", "END", "ignored"])
count = 0
while (line := next(data)) != "END":
    if line != "":
        count += 1
print("Non-empty lines before END:", count)
```

## Summary and Key Points

- `if`/`elif`/`else` branches on conditions in order; the conditional expression `a if cond else b` handles simple cases on one line.
- `for` iterates over sequences; `while` repeats on a condition. `range()`, `enumerate()`, and `zip()` are everyday tools.
- `break` exits a loop, `continue` skips an iteration, `pass` is a placeholder, and a loop `else` runs only when no `break` occurred.
- `match`/`case` matches literals, sequences, mappings, and class shapes, supports OR patterns and guards, and uses `_` as the default.
- The walrus operator `:=` assigns inside an expression, simplifying read-and-test loops and comprehensions.
