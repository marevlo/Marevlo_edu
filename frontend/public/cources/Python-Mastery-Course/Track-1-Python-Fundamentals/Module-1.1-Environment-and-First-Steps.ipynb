## Part 1 · The Three Ways to Run Python

Python code can be executed in three different environments. All three run the same language; the difference lies in how code is supplied and how results are returned.

| Environment | Description | Best suited for |
|---|---|---|
| Interactive interpreter (REPL) | A live prompt (`>>>`) where you type one line and immediately see the result | Quick experiments and checking behaviour |
| `.py` script file | A text file of code that is executed in full | Real programs, automation, anything reused |
| Notebook (the current environment) | Code divided into runnable cells, interleaved with explanatory text | Learning, data analysis, documented workflows |

Each environment is examined below.

### 1a · The interactive interpreter (the REPL)

REPL stands for Read, Evaluate, Print, Loop. You open a terminal, type `python`, and are presented with a `>>>` prompt. You enter one expression, press Enter, and Python immediately displays the result before waiting for the next line.

```text
$ python
>>> 2 + 2
4
>>> "hi" * 3
'hihihi'
```

It functions as a calculator that understands Python, and is ideal for testing a small idea. Its limitation is that nothing is saved: once the session is closed, the work is lost. This is the reason we write scripts.

### 1b · Running a `.py` script file

A script is a plain text file ending in `.py` that contains Python code. It is executed in full from a terminal with the command `python myfile.py`.

A notebook cannot fully reproduce a terminal session; that is demonstrated in the short setup recording accompanying this module. However, a notebook is able to create a `.py` file and run it, which lets us observe the script model directly. Run the next two cells.

````python
%%writefile hello.py
# The %%writefile command (which must be the first line of the cell) saves everything
# below it into a file named hello.py instead of executing it.
# The result is a complete Python program in its own file.
print("Hello from a real .py script.")
print("Scripts run from top to bottom, all at once.")
````

````python
# A leading "!" runs a terminal command from within the notebook.
# Here we instruct Python to run the file just created, exactly as in a terminal.
!python hello.py
````

The file ran from start to finish in a single step and printed both lines. This is the script model: the complete set of instructions is written first, then executed as a unit. Nearly all production Python software is delivered this way.

### 1c · Notebooks: cells, the kernel, and state

A notebook is composed of cells. Each code cell is executed independently when you press `Shift + Enter`. Behind the notebook runs a kernel, a single Python process that remains active and retains everything you have run.

This retained information is called state, and it is the most important concept to understand about notebooks:

- A variable created in one cell remains available in later cells.
- Cells execute in the order you run them, not in the order they appear on the page. The number shown beside a cell, such as `[5]`, indicates it was the fifth cell executed.

The following two cells demonstrate this.

````python
# Cell A: a value is created here.
favorite_number = 7
print("The value", favorite_number, "is now stored in the kernel's memory.")
````

````python
# Cell B: a separate cell can still access it, because the kernel retains state.
print("Two cells later, the kernel still has the value:", favorite_number)
print("It can be used in a calculation:", favorite_number * 6)
````

> **Caution — a common beginner error.** Because cells retain state and may be run out of order, it is possible to mislead yourself. If you define something, delete the cell, but never restart the kernel, the value remains in memory. The notebook then works for you but fails for anyone who runs it from a fresh state.
>
> **Recommended practice.** Before sharing a notebook, or whenever the behaviour seems inconsistent, use the menu option **Kernel → Restart & Run All**. This clears the kernel's memory and runs every cell from the top, reproducing the experience of a new reader. Adopting this habit early is worthwhile.

## Part 2 · `print()` in Depth

`print("Hello")` is usually the first line a learner writes, and it is easy to assume there is nothing more to it. In fact `print()` has several parameters that make it considerably more capable.

Its full signature is:

```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

In plain terms:

- `*objects` — any number of values may be passed, separated by commas. Python converts each one to text automatically.
- `sep=' '` — the separator placed between those values. The default is a single space.
- `end='\n'` — the text printed after all values. The default is `\n`, a newline, which is why each call to `print` begins on a new line.
- `file` — the destination of the output. The default is the screen (`stdout`).
- `flush` — whether to force the text out immediately, which matters for live progress indicators.

The cells below illustrate each parameter.

````python
# Several values of different types may be passed; print converts them all to text.
print("Age:", 30, "Height:", 1.75, "Student?", True)
````

````python
# sep controls the text placed BETWEEN values, which is convenient for formatting.
print("2025", "05", "30", sep="/")     # produces 2025/05/30
print("a", "b", "c", sep=" -> ")        # produces a -> b -> c
print("no", "spaces", sep="")           # produces nospaces
````

````python
# end controls the text placed AFTER the output. Setting end="" keeps the next
# print on the SAME line.
print("Loading", end="")
print("...", end="")
print(" complete.")   # the three calls appear on one line: Loading... complete.
````

````python
import sys

# The file parameter directs output to the error stream (stderr) rather than the
# standard output stream (stdout). Programs use this to separate results from
# warnings and log messages.
print("This is a normal result.")                    # sent to stdout
print("This is a warning message.", file=sys.stderr)  # sent to stderr
````

### A notebook-specific behaviour: automatic display versus `print()`

In a notebook, if the final line of a cell is simply a value, the notebook displays it automatically without `print()`. This is a convenience provided by the notebook, not a feature of the Python language; a `.py` script would display nothing. Compare the following two cells.

````python
3 + 4        # not explicitly printed, yet the notebook displays the final value
````

````python
10 + 1       # only the LAST bare value is displayed automatically
20 + 2       # this one is displayed
print("Use print() whenever output must be guaranteed, particularly in scripts.")
````

**Conclusion:** automatic display is useful for quick inspection while exploring, but `print()` should be used whenever output must be reliable. This is the practice that transfers to production programs.

## Part 3 · Comments and Docstrings

Code instructs the computer what to do. Comments and docstrings explain to people why it does so. One of those people is you, several months from now, which is why this matters more than beginners often expect.

### Comments — `#`

Everything after a `#` on a line is ignored by Python. Comments should explain the reasoning, not restate the obvious.

````python
# Useful comment: explains the reasoning, which is not obvious from the code.
price = 100 * 1.18   # apply 18% GST to obtain the final customer price

# Unhelpful comment: merely repeats what the code already states.
total = price + 5    # add 5 to price

print("Final total:", total)
````

### Docstrings — `"""..."""`

A docstring is a string placed as the first statement inside a function, class, or module. Unlike a comment, Python retains it and attaches it to the object as documentation. It is the standard means of describing what something does, as set out in PEP 257.

Two capabilities follow automatically:

- `help(object)` displays the docstring as formatted documentation.
- `object.__doc__` returns the raw docstring text.

This is precisely how `help(print)` works; `print` itself has a docstring.

````python
def greet(name):
    """Return a friendly greeting for the given name.

    The text enclosed in triple quotes is a docstring: it resides inside the
    function and documents it.
    """
    return "Hello, " + name + "."

print(greet("Ada"))           # call the function normally
print("---")
print(greet.__doc__)          # the raw docstring text
print("---")
help(greet)                   # the same docstring, formatted as help
````

Run `help(print)` or `help(len)` in a new cell. The text displayed is the docstring written by the authors of Python. Reading documentation in this manner is an essential professional skill.

### Example 2 — Building a date string with `sep` (Easy)

Demonstrates control of the separator between values.

````python
# Rather than joining strings manually, allow the sep parameter to do the work.
year, month, day = 2025, 5, 30

print(year, month, day, sep="-")   # ISO style:    2025-5-30
print(day, month, year, sep="/")   # common style: 30/5/2025

# Experiment: change sep to "." or " " and observe the change in format.
````

### Example 3 — A multi-line information card (Medium)

Demonstrates escape characters. `\n` denotes a new line and `\t` denotes a tab, which together allow a single `print` to produce neatly arranged text.

````python
# \n begins a new line; \t inserts a tab for alignment.
card = "Name:\tAda Lovelace\nRole:\tProgrammer\nNote:\tWrote the first algorithm"

print("===== INFORMATION CARD =====")
print(card)
print("============================")

# Experiment: add a further line, for example "\nCity:\tLondon", and run the cell again.
````

### Example 4 — A self-documenting function (Medium)

Demonstrates docstrings together with `help()`.

````python
def area_of_rectangle(width, height):
    """Compute the area of a rectangle.

    Args:
        width:  the width of the rectangle (a number)
        height: the height of the rectangle (a number)
    Returns:
        The area, equal to width multiplied by height.
    """
    return width * height

# Use the function, then read its documentation.
print("Area:", area_of_rectangle(4, 5))
help(area_of_rectangle)

# Experiment: revise the docstring, run the cell again, and observe how help() changes.
````

### Example 5 — An aligned receipt (Difficult)

Combines `sep`, `end`, and basic string operations. Note how `end=""` keeps output on one line and how `"." * n` repeats a character `n` times.

````python
items = [("Coffee", 120), ("Sandwich", 90), ("Cookie", 45)]

print("RECEIPT")
print("-" * 24)               # a divider consisting of 24 dashes
for item, price in items:
    # Print the name, then enough dots to pad to column 16, then the price.
    print(item, end="")
    dots = "." * (16 - len(item))
    print(dots, end="")
    print("Rs", price)
print("-" * 24)

# Experiment: add ("Tea", 30) to the items list and run the cell again.
````

### Example 6 — A countdown updated in place (Difficult)

This is the mechanism underlying progress indicators. By printing with `end="\r"` (a carriage return, which moves the cursor to the start of the line) and `flush=True` (which sends the output immediately), each message overwrites the previous one rather than scrolling.

````python
import time

for n in range(5, 0, -1):                 # 5, 4, 3, 2, 1
    print(f"Launching in {n}... ", end="\r", flush=True)
    time.sleep(0.6)                       # pause so each number is visible

print("Launch complete.          ")       # trailing spaces clear leftover characters

# Experiment: change range(5, 0, -1) to range(10, 0, -1) for a longer countdown.
````

## Summary and Key Points

- Three runtimes exist: the REPL for live experimentation, `.py` scripts for complete programs run as a unit, and notebooks combining cells and text for learning and analysis.
- The kernel retains state, and cells execute in the order they are run rather than the order shown. When behaviour is unclear, use Kernel → Restart & Run All.
- `print()` is more capable than it first appears: `sep` controls text between values, `end` controls text after them (`end=""` stays on one line, `end="\r"` overwrites the line), `file` selects stdout or stderr, and `flush` forces output to appear immediately.
- Comments (`#`) explain reasoning, while docstrings (`"""..."""`) document functions and provide the content for `help()` and `__doc__`.

**Points to remember**

- Executing cells out of order can make a flawed notebook appear correct. Restart and run all cells before sharing.
- Automatic display of a final bare value is a notebook convenience; scripts do not do this. Use `print()` when output must be guaranteed.
