## Part 1 · Variables and Dynamic Typing

A variable is a name that refers to a value. You create one with the assignment operator `=`, writing the name on the left and the value on the right.

A useful mental model: a variable is a **label** attached to a value, not a box that contains it. Assigning a new value simply moves the label to a different value. Because of this, Python is **dynamically typed**: a name is not tied to one type and may refer to a value of any type over its lifetime.

```python
x = 10          # x now labels an integer
print(x, type(x))

x = "ten"       # the same name now labels a string; this is perfectly legal
print(x, type(x))

# type() reports the type of any value. It is your primary tool for inspecting data.
```

```python
# Several names can be assigned at once.
a, b, c = 1, 2, 3
print(a, b, c)

# A single value can be given to several names at once.
p = q = 0
print(p, q)

# Names follow rules: letters, digits, and underscores; they may not start with a digit,
# and should be lowercase_with_underscores by convention (see PEP 8).
total_score = 95
print(total_score)
```

## Part 2 · The Core Types

Five types account for the vast majority of everyday Python. Each has its own literal syntax.

| Type | Meaning | Example literal |
|---|---|---|
| `int` | whole number, unlimited size | `42`, `-7`, `1_000_000` |
| `float` | number with a fractional part | `3.14`, `-0.5`, `2.0` |
| `str` | text (a sequence of characters) | `"hello"`, `'a'` |
| `bool` | truth value | `True`, `False` |
| `NoneType` | the absence of a value | `None` |

Two details worth noting early: integers in Python have no fixed size limit, and `bool` is technically a subtype of `int`, so `True` equals `1` and `False` equals `0`.

```python
big = 2 ** 100          # integers can be arbitrarily large
print("A very large integer:", big)

ratio = 7 / 2           # division always produces a float
print("Type of 7 / 2:", type(ratio), "value:", ratio)

print("True + True =", True + True)   # bool behaves like int here, giving 2

nothing = None          # None represents 'no value'; it is not zero and not ""
print("nothing is:", nothing, type(nothing))
```

Underscores may be used as digit separators (`1_000_000`) to make large numbers readable; Python ignores them. Floats are stored in binary and cannot represent every decimal exactly, which leads to small rounding effects you should be aware of.

```python
print(1_000_000)               # readable large number

# A famous example of floating-point imprecision:
print(0.1 + 0.2)               # not exactly 0.3
print(0.1 + 0.2 == 0.3)        # therefore this is False
# For exact decimal arithmetic (for example, money), the decimal module is used later.
```

## Part 3 · Type Conversion

Values can be converted from one type to another using the type names as functions: `int()`, `float()`, `str()`, and `bool()`. This is essential because input often arrives as text that must be turned into numbers.

```python
age_text = "30"                 # text, perhaps from a form
age = int(age_text)             # convert to an integer
print(age + 1, type(age))

print(float("3.14"))            # text to float
print(str(2025))                # number to text, useful for joining with other text
print(int(3.99))                # float to int TRUNCATES toward zero; it does not round
```

```python
# Not every conversion succeeds. int() requires a clean integer string.
# int("3.5") would raise an error because "3.5" is not an integer.
# We handle such failures with try/except in Module 1.5.

# bool() conversion follows truthiness rules (see Part 5):
print(bool(0), bool(1), bool(""), bool("a"), bool([]), bool([1]))
```

## Part 4 · Operators

Operators combine values to produce new ones. They fall into several families.

**Arithmetic:** `+`, `-`, `*`, `/` (true division, always a float), `//` (floor division), `%` (remainder), `**` (power).

**Comparison:** `==`, `!=`, `<`, `>`, `<=`, `>=`. These produce a `bool`.

**Logical:** `and`, `or`, `not`. They combine boolean values and use short-circuit evaluation.

**Membership:** `in`, `not in`. They test whether a value occurs within a container.

**Identity:** `is`, `is not`. They test whether two names refer to the **same object**, which is different from testing equality.

```python
print("Arithmetic")
print(7 / 2)     # 3.5  (true division)
print(7 // 2)    # 3    (floor division: drops the fractional part)
print(7 % 2)     # 1    (remainder)
print(2 ** 10)   # 1024 (power)
```

```python
print("Comparison and logical")
print(5 > 3, 5 == 5, 5 != 4)
print(True and False, True or False, not True)

# Short-circuit: 'and' stops at the first false value; 'or' stops at the first true value.
# The right-hand side below is never evaluated because the left of 'or' is already True.
print(True or (1 / 0 == 0))
```

```python
print("Membership and identity")
print("a" in "cat", "z" not in "cat")     # membership works on strings and containers

# Identity versus equality: == compares values; is compares object identity.
list_one = [1, 2, 3]
list_two = [1, 2, 3]
print(list_one == list_two)   # True: same contents
print(list_one is list_two)   # False: two different objects in memory
# Rule of thumb: use 'is' only for None (for example, 'if x is None'), and == for values.
```

## Part 5 · Precedence and Truthiness

When several operators appear together, **precedence** decides the order of evaluation. Power binds tightest, then unary minus, then `*` `/` `//` `%`, then `+` `-`, then comparisons, then `not`, then `and`, then `or`. When in doubt, add parentheses; clarity is worth more than cleverness.

**Truthiness** is the idea that every value can be treated as true or false in a boolean context. The values considered false are: `False`, `None`, zero of any numeric type, and every empty container (`""`, `[]`, `{}`, `()`). Everything else is true.

```python
# Precedence: ** before *, * before +, comparison after arithmetic.
print(2 + 3 * 4)        # 14, not 20
print(2 ** 3 ** 2)      # 512, because ** is right-associative: 2 ** (3 ** 2)
print((2 + 3) * 4)      # 20, parentheses make intent explicit

# Truthiness in action:
name = ""
print("Has a name" if name else "Name is empty")   # empty string is falsy
```

### Example 2 — Reading numbers from text (Easy)

```python
a_text = "12"
b_text = "8"

# Without conversion, "+" would join the text as "128". With conversion, we add numbers.
print("As text:", a_text + b_text)
print("As numbers:", int(a_text) + int(b_text))

# Experiment: change the values, and try making one of them a decimal with float().
```

### Example 3 — Even, odd, and divisibility (Medium)

Uses the remainder operator and comparison.

```python
number = 18
print(number, "is even" if number % 2 == 0 else "is odd")
print("Divisible by 3:", number % 3 == 0)
print("Quotient and remainder when divided by 5:", number // 5, "remainder", number % 5)

# Experiment: change number to an odd value such as 17.
```

### Example 4 — An eligibility check with logical operators (Medium)

Uses `and`, `or`, `not`, and comparison.

```python
age = 20
is_citizen = True
has_id = False

can_vote = age >= 18 and is_citizen
print("May vote:", can_vote)

needs_document = not has_id
print("Must bring identification:", needs_document)

# A combined condition: entry is allowed if a member, or an adult with a guest pass.
is_member = False
has_guest_pass = True
allowed = is_member or (age >= 18 and has_guest_pass)
print("Entry allowed:", allowed)

# Experiment: set is_citizen to False and see how can_vote changes.
```

### Example 5 — Identity versus equality (Difficult)

Demonstrates the difference between `is` and `==`, a frequent source of bugs.

```python
x = [1, 2, 3]
y = x              # y is the SAME list as x (same label on the same object)
z = list(x)        # z is a NEW list with equal contents

print("x == y:", x == y, " x is y:", x is y)   # equal and identical
print("x == z:", x == z, " x is z:", x is z)   # equal but NOT identical

x.append(4)        # because y is the same object, changing x also changes y
print("After x.append(4):  y =", y, " z =", z)

# Conclusion: use 'is' for identity (notably 'is None'); use '==' to compare values.
```

### Example 6 — Precedence and truthiness combined (Difficult)

Predict each result before running, then check.

```python
print(10 - 2 ** 3)              # ** first: 10 - 8 = 2
print(not 0)                    # 0 is falsy, so not 0 is True
print(1 + 2 > 2 and 3 < 1)      # arithmetic, then comparison, then and: True and False
print("" or "fallback")         # or returns the first truthy value: "fallback"
print(5 and 0 and 9)            # and returns the first falsy value (or the last): 0

# Experiment: rewrite the third line with explicit parentheses to confirm the order.
```

## Summary and Key Points

- A variable is a name (label) bound to a value; Python is dynamically typed, so a name may refer to any type. Use `type()` to inspect values.
- The core types are `int`, `float`, `str`, `bool`, and `None`. Integers are unbounded, division yields a float, and floats carry small binary rounding effects.
- Convert with `int()`, `float()`, `str()`, `bool()`; conversions can fail (handled in Module 1.5), and `int()` of a float truncates rather than rounds.
- Operators fall into arithmetic, comparison, logical, membership, and identity families. Reserve `is` for identity (especially `is None`) and use `==` for value comparison.
- Precedence determines evaluation order; use parentheses for clarity. Falsy values are `False`, `None`, zero, and empty containers; all else is truthy.
