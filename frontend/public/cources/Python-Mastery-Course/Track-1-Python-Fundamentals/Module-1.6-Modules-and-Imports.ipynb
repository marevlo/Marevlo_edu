## Part 1 · The Import Statements

`import name` loads a module and makes its contents available under that name, accessed with a dot, as in `math.sqrt`. `from name import thing` brings a specific name directly into your namespace. `as` assigns an alias, which is useful for long names or community conventions.

Importing a module runs it once; Python caches it, so later imports are fast and do not re-run the code.

```python
import math                      # whole-module import
print(math.sqrt(144), math.pi)

from random import randint       # import one specific function
print(randint(1, 6))             # a value from 1 to 6 inclusive

import statistics as stats       # import under an alias
print(stats.mean([2, 4, 6]))
```

A common convention is `from module import name` for a few frequently used names, and `import module` (or an alias) otherwise. Avoid `from module import *`, which imports everything and makes it unclear where names came from.

## Part 2 · Creating Your Own Module

Any `.py` file is a module. The cell below writes a small file named `geometry.py`, and the following cell imports and uses it, exactly as you would in a real project.

```python
%%writefile geometry.py
"""A tiny geometry module for demonstration."""

PI = 3.141592653589793

def circle_area(radius):
    """Return the area of a circle with the given radius."""
    return PI * radius * radius

def rectangle_area(width, height):
    """Return the area of a rectangle."""
    return width * height
```

```python
import geometry                       # imports the file we just wrote

print("PI from module:", geometry.PI)
print("circle area:", geometry.circle_area(2))

from geometry import rectangle_area  # bring in just one function
print("rectangle area:", rectangle_area(3, 4))
```

## Part 3 · The `__name__ == "__main__"` Idiom

Every module has a built-in variable `__name__`. When a file is **run directly**, Python sets its `__name__` to the string `"__main__"`. When the same file is **imported**, `__name__` is set to the module's name instead.

This lets a file act both as a reusable module and as a runnable script: code guarded by `if __name__ == "__main__":` runs only when the file is executed directly, not when it is imported. The next cells demonstrate both behaviours.

```python
%%writefile tool.py
"""A module that can also be run as a script."""

def add(a, b):
    return a + b

print("This top-level line runs on BOTH import and direct run. __name__ =", __name__)

if __name__ == "__main__":
    # This block runs ONLY when the file is executed directly.
    print("Running as a script: add(2, 3) =", add(2, 3))
```

```python
# Run the file directly, as if from a terminal. __name__ becomes "__main__",
# so the guarded block runs.
!python tool.py
```

```python
# Now IMPORT the same file. __name__ becomes "tool", so the guarded block is skipped;
# only the top-level line runs, and the function becomes available to us.
import tool
print("Imported. tool.add(10, 5) =", tool.add(10, 5))
```

## Part 4 · A Tour of the Standard Library

Python ships with a large standard library, so much functionality needs no installation. Four modules you will reach for constantly:

- `math` — mathematical functions and constants.
- `random` — random numbers and choices (seed it for reproducible results).
- `sys` — interpreter and runtime information.
- `os` — interaction with the operating system and file system.

```python
import math
print("floor/ceil of 3.4:", math.floor(3.4), math.ceil(3.4))
print("factorial of 5:", math.factorial(5))
print("gcd of 24 and 36:", math.gcd(24, 36))
```

```python
import random
random.seed(42)                  # fixing the seed makes results reproducible
print("random int 1..100:", random.randint(1, 100))
print("random choice:", random.choice(["red", "green", "blue"]))
deck = [1, 2, 3, 4, 5]
random.shuffle(deck)             # shuffles in place
print("shuffled:", deck)
```

```python
import sys, os
print("Python version:", sys.version.split()[0])
print("current working directory:", os.getcwd())
# os.listdir shows files in a directory; we filter to the .py files we created.
py_files = [f for f in os.listdir(".") if f.endswith(".py")]
print("python files here:", sorted(py_files))
```

## Part 5 · Packages, Briefly

When a project grows beyond one file, modules are grouped into a **package**: a directory of modules. Importing then uses dotted paths such as `import mypackage.submodule` or `from mypackage import submodule`. Modern Python treats most directories as packages automatically; historically a directory needed an `__init__.py` file to mark it as one, and that file is still used to control what a package exposes. Packaging for distribution is covered in Track 6.

```python
from math import sqrt, pi as PI       # import two names, one aliased
import statistics as stats

data = [4, 8, 15, 16, 23, 42]
print("mean:", stats.mean(data))
print("standard deviation:", round(stats.pstdev(data), 3))
print("sqrt of len:", sqrt(len(data)), "PI:", PI)
```

### Example 4 — Writing and importing a custom module (Medium)

```python
%%writefile temperature.py
"""Conversions between Celsius and Fahrenheit."""

def c_to_f(c):
    return c * 9 / 5 + 32

def f_to_c(f):
    return (f - 32) * 5 / 9
```

```python
import temperature
print("100 C in F:", temperature.c_to_f(100))
print("212 F in C:", temperature.f_to_c(212))
```

### Example 5 — Demonstrating `__name__` both ways (Difficult)

```python
%%writefile report.py
"""Generates a report; usable as a module or a script."""

def make_report(title, items):
    lines = [title, "-" * len(title)]
    lines += [f"- {item}" for item in items]
    return "\n".join(lines)

if __name__ == "__main__":
    print(make_report("Daily Tasks", ["plan", "build", "review"]))
```

```python
print("=== run directly (guard runs) ===")
!python report.py

print("\n=== imported (guard skipped, function reused) ===")
import report
print(report.make_report("Imported Use", ["a", "b"]))
```

## Summary and Key Points

- `import module` accesses names with a dot; `from module import name` brings names in directly; `as` provides an alias. Avoid `import *`.
- Any `.py` file is a module. Importing it runs the file once and caches it.
- `__name__` equals `"__main__"` when a file is run directly and the module's name when imported; guard script-only code with `if __name__ == "__main__":`.
- The standard library is extensive; `math`, `random` (seed it for reproducibility), `sys`, and `os` are everyday staples.
- Packages group modules into directories, imported via dotted paths; distribution packaging is covered later in Track 6.
