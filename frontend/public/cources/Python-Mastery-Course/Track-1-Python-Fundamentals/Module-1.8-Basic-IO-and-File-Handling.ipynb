## Part 2 · Opening, Reading, and Writing Files

`open(path, mode)` connects your program to a file. The mode selects the operation:

| Mode | Meaning |
|---|---|
| `"r"` | read (the default); error if the file is missing |
| `"w"` | write; creates the file or **replaces** its contents |
| `"a"` | append; adds to the end, keeping existing contents |
| `"r+"` | read and write |
| adding `"b"` | binary mode, for non-text data such as images |

Reading offers several approaches: `read()` returns the whole file as one string, `readline()` returns one line, `readlines()` returns a list of lines, and iterating over the file yields lines one at a time, which is memory-efficient for large files.

```python
# Write three lines to a file. "w" creates or replaces it.
file = open("sample.txt", "w")
file.write("first line\n")        # we add the newline ourselves
file.write("second line\n")
file.write("third line\n")
file.close()                       # closing flushes and releases the file

# Read the whole file back as one string.
file = open("sample.txt", "r")
contents = file.read()
file.close()
print(contents)
```

## Part 3 · The `with` Statement

The examples above had to remember to call `close()`. If an error occurred in between, the file might never be closed. The `with` statement solves this: it **automatically closes** the file when the block ends, even if an exception is raised. This is the recommended way to work with files.

```python
# Writing with 'with': no explicit close needed.
with open("sample.txt", "w") as f:
    f.write("apple\n")
    f.write("banana\n")
    f.write("cherry\n")
# The file is closed automatically here, at the end of the block.

# Reading line by line by iterating over the file object (memory-efficient):
with open("sample.txt", "r") as f:
    for line_number, line in enumerate(f, start=1):
        print(line_number, line.strip())   # strip removes the trailing newline
```

```python
# The different read methods compared:
with open("sample.txt", "r") as f:
    print("read() ->", repr(f.read()))        # whole file as one string

with open("sample.txt", "r") as f:
    print("readlines() ->", f.readlines())    # list of lines, newlines kept

# Append mode adds without erasing:
with open("sample.txt", "a") as f:
    f.write("date\n")

with open("sample.txt", "r") as f:
    print("after append ->", f.read().split())
```

## Part 4 · File Paths with `pathlib`

Building paths by joining strings is error-prone across operating systems. The modern `pathlib` module represents a path as a `Path` object with convenient methods and the `/` operator for joining. It also offers shortcuts `read_text()` and `write_text()` for simple whole-file access.

```python
from pathlib import Path

path = Path("notes.txt")               # a path relative to the current folder

path.write_text("remember to practise\n")   # write the whole file in one call
print("exists:", path.exists())
print("name:", path.name, "| suffix:", path.suffix)   # notes.txt | .txt
print("contents:", repr(path.read_text()))

# Joining paths with the / operator, and listing files by pattern:
folder = Path(".")
text_files = sorted(folder.glob("*.txt"))    # all .txt files in this folder
print("text files here:", [p.name for p in text_files])
```

## Part 5 · Robust File Access

Files may be missing or unreadable. Combining `with` and `try`/`except` (from Module 1.5) produces programs that fail gracefully. The most common case to handle is `FileNotFoundError`.

```python
def read_safely(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        return f"(no file named {filename})"

print(read_safely("sample.txt")[:20], "...")
print(read_safely("does_not_exist.txt"))
```

### Example 3 — Counting lines and words in a file (Medium)

```python
with open("article.txt", "w") as f:
    f.write("Python is clear.\n")
    f.write("Python is powerful.\n")
    f.write("Practice makes progress.\n")

line_count = 0
word_count = 0
with open("article.txt", "r") as f:
    for line in f:                       # iterate line by line
        line_count += 1
        word_count += len(line.split())

print("lines:", line_count, "words:", word_count)
```

### Example 4 — Appending and reviewing a log (Medium)

```python
log = "events.log"

# Start fresh, then append several entries.
with open(log, "w") as f:
    f.write("start\n")
for event in ["login", "download", "logout"]:
    with open(log, "a") as f:           # append keeps previous lines
        f.write(event + "\n")

with open(log, "r") as f:
    entries = [line.strip() for line in f]
print("log entries:", entries)
```

### Example 5 — A pathlib workflow (Difficult)

```python
from pathlib import Path

folder = Path("data_demo")
folder.mkdir(exist_ok=True)             # create the folder if it is not there

# Write three small files inside it.
for i in range(1, 4):
    (folder / f"file{i}.txt").write_text(f"contents of file {i}\n")

# List them, read each, and report sizes.
for path in sorted(folder.glob("*.txt")):
    text = path.read_text()
    print(f"{path.name}: {len(text)} characters -> {text.strip()}")
```

### Example 6 — Robust reading with error handling (Difficult)

```python
from pathlib import Path

def summarise(filename):
    path = Path(filename)
    try:
        text = path.read_text()
    except FileNotFoundError:
        return f"{filename}: not found"
    words = text.split()
    return f"{filename}: {len(text.splitlines())} lines, {len(words)} words"

print(summarise("article.txt"))         # exists from Example 3
print(summarise("missing.txt"))         # handled gracefully
```

## Summary and Key Points

- `input()` returns text; convert it with `int()` or `float()` when a number is needed.
- `open(path, mode)` selects the operation: `"r"` read, `"w"` write (replaces), `"a"` append, and `"b"` for binary. Read with `read()`, `readline()`, `readlines()`, or by iterating the file.
- The `with` statement closes files automatically, even on error, and is the recommended approach.
- `pathlib.Path` represents paths cleanly, joins them with `/`, and offers `read_text()`, `write_text()`, `exists()`, and `glob()`.
- Combine `with` and `try`/`except` (notably `FileNotFoundError`) for robust file access.
