## Part 1 · String Basics: Creation, Indexing, Slicing

A string is a sequence of characters, written in single or double quotes (they are equivalent). Triple quotes allow text spanning several lines. Strings are **immutable**: methods that appear to change a string actually return a new one.

Each character has an index starting at 0; negative indices count from the end. **Slicing**, written `text[start:stop:step]`, extracts a substring up to but not including `stop`.

```python
text = "Python"
print("length:", len(text))
print("first character:", text[0])     # P
print("last character:", text[-1])     # n

print("slice [0:3]:", text[0:3])       # Pyt  (stop index 3 is excluded)
print("slice [2:]:", text[2:])         # thon
print("every second char:", text[::2]) # Pto
print("reversed:", text[::-1])         # nohtyP

# Immutability: this would raise an error, because strings cannot be modified in place:
# text[0] = "J"
new_text = "J" + text[1:]              # instead, build a new string
print("rebuilt:", new_text)
```

Concatenation uses `+`, and repetition uses `*`. Multi-line strings use triple quotes.

```python
print("abc" + "def")
print("ha" * 3)

paragraph = """Line one.
Line two.
Line three."""
print(paragraph)
```

## Part 2 · Common String Methods

Strings carry many helpful methods. The most frequently used clean, search, split, and join text. Because strings are immutable, each method returns a new string.

```python
raw = "   Hello, World   "
print("strip:", repr(raw.strip()))           # remove surrounding whitespace
print("lower:", raw.strip().lower())
print("upper:", raw.strip().upper())
print("replace:", raw.strip().replace("World", "Python"))
print("startswith Hello:", raw.strip().startswith("Hello"))
print("find 'World':", raw.find("World"))    # index of first match, or -1
```

```python
# split breaks a string into a list; join combines a list into a string.
csv_line = "apple,banana,cherry"
parts = csv_line.split(",")
print("parts:", parts)
print("rejoined with ' | ':", " | ".join(parts))

sentence = "the quick brown fox"
print("title case:", sentence.title())
print("word count:", len(sentence.split()))
```

## Part 3 · f-strings and the Format Specification Mini-Language

An **f-string** is a string prefixed with `f`, in which expressions inside braces `{}` are evaluated and inserted. It is the clearest and fastest way to build text in modern Python.

After a colon inside the braces comes the **format specifier**, which controls width, alignment, precision, and more. A few essentials:

- `:.2f` shows a float with two decimal places.
- `:,` inserts thousands separators.
- `:>10`, `:<10`, `:^10` right-, left-, and centre-align within a width of 10.
- `:.1%` formats a number as a percentage.
- The `{value=}` form (Python 3.8+) prints both the expression and its value, which is excellent for debugging.

```python
name = "Asha"
score = 0.8736
amount = 1234567

print(f"Hello, {name}. Welcome.")
print(f"Score: {score:.2f}")            # 0.87
print(f"As percent: {score:.1%}")       # 87.4%
print(f"Amount: {amount:,}")            # 1,234,567
print(f"Padded: '{name:>10}'")          # right-aligned in 10 columns
print(f"Centred: '{name:^10}'")         # centred in 10 columns

# The = form shows the expression and its value together; ideal for quick debugging.
x = 42
print(f"{x = }")                        # x = 42
print(f"{x * 2 = }")                    # x * 2 = 84
```

## Part 4 · Older Formatting Styles

Before f-strings, Python used `str.format()` and the `%` operator. You will meet both in existing code, so it helps to recognise them, but f-strings are preferred for new code.

```python
name, age = "Ben", 30

# str.format() uses positional or named placeholders.
print("{} is {} years old".format(name, age))
print("{n} is {a}".format(n=name, a=age))

# The older % style, common in legacy code and some logging.
print("%s is %d years old" % (name, age))
```

## Part 5 · Text Versus Bytes

A `str` is a sequence of **characters**; `bytes` is a sequence of **raw bytes** (numbers 0 to 255). Files and networks deal in bytes, so text must be **encoded** to bytes when sent and **decoded** back when received. The standard encoding is UTF-8, which represents every character but uses more than one byte for many non-ASCII characters.

This is why the number of characters and the number of bytes can differ.

```python
word = "café"          # the accented e is a non-ASCII character

encoded = word.encode("utf-8")     # str -> bytes
print("type:", type(encoded), "value:", encoded)
print("characters:", len(word), "bytes:", len(encoded))   # 4 characters, 5 bytes

decoded = encoded.decode("utf-8")  # bytes -> str
print("decoded back:", decoded, "equal to original:", decoded == word)

# A character outside the Latin alphabet uses even more bytes:
symbol = "東"           # a Japanese character
print("'東' encodes to", len(symbol.encode("utf-8")), "bytes")
```

The `textwrap` module reflows long text to a given width, which is useful for tidy console output.

```python
import textwrap
long_text = ("Python strings are immutable sequences of characters, and Python "
             "provides a rich set of methods and a powerful formatting system.")
print(textwrap.fill(long_text, width=50))
```

### Example 1 — Cleaning and reformatting a name (Easy)

```python
raw = "  jOHN   smith  "
clean = raw.strip().title()      # remove spaces, capitalise each word
print(repr(clean))               # 'John   Smith' (inner spaces remain)
# Collapse inner spaces by splitting and rejoining:
print(" ".join(clean.split()))   # 'John Smith'
# Experiment: try a different messy name.
```

### Example 2 — Formatting numbers with f-strings (Easy)

```python
price = 2499.5
quantity = 12000

print(f"Price: {price:.2f}")          # two decimals
print(f"Quantity: {quantity:,}")      # thousands separator
print(f"Total: {price * 3:,.2f}")     # both at once
# Experiment: change the precision to .3f or add alignment such as :>12.2f.
```

### Example 3 — Parsing and transforming a line (Medium)

```python
line = "name=Asha;city=Pune;role=engineer"

pairs = line.split(";")                        # ['name=Asha', 'city=Pune', ...]
record = {}
for pair in pairs:
    key, value = pair.split("=")               # split each into key and value
    record[key] = value

print(record)
print(f"{record['name']} from {record['city']} works as a(n) {record['role']}.")
```

### Example 4 — An aligned table with format specifiers (Medium)

```python
rows = [("Apples", 3, 45.0), ("Bread", 2, 30.5), ("Milk", 1, 60.0)]

# Column widths and alignment make a tidy table. < left, > right, ^ centre.
print(f"{'Item':<10}{'Qty':>5}{'Price':>10}")
print("-" * 25)
for item, qty, price in rows:
    print(f"{item:<10}{qty:>5}{price:>10.2f}")
# Experiment: adjust the widths and watch the columns shift.
```

### Example 5 — A wrapped, formatted report (Difficult)

```python
import textwrap

title = "Quarterly Summary"
body = ("Revenue grew steadily across all regions, with the strongest results "
        "in the southern market. Costs remained controlled, and the outlook "
        "for the next quarter is cautiously positive.")
figure = 0.142

report = []
report.append(title)
report.append("=" * len(title))
report.append(textwrap.fill(body, width=55))
report.append("")
report.append(f"Growth rate: {figure:.1%}")

print("\n".join(report))
```

### Example 6 — Text, bytes, and encoding (Difficult)

```python
messages = ["hello", "café", "東京", "naïve"]

for m in messages:
    b = m.encode("utf-8")
    print(f"{m:<8} chars={len(m):<2} bytes={len(b):<2} encoded={b}")

# Decoding with the wrong assumption can fail; UTF-8 round-trips correctly:
data = "Zürich".encode("utf-8")
print("round trip:", data.decode("utf-8"))
```

## Summary and Key Points

- Strings are immutable sequences of characters; index with `[]`, extract with slices `[start:stop:step]`, and build new strings with `+` and `*`.
- Common methods (`strip`, `lower`, `upper`, `replace`, `split`, `join`, `find`, `startswith`, `title`) clean and transform text and always return new strings.
- f-strings insert expressions in braces and accept a format specifier after a colon for precision (`.2f`), thousands separators (`,`), alignment (`<`, `>`, `^`), and percentages (`%`); the `{value=}` form aids debugging.
- `str.format()` and `%` are older styles you will see in existing code, but f-strings are preferred.
- `str` is characters and `bytes` is raw bytes; encode to send and decode to receive, using UTF-8, where non-ASCII characters take more than one byte.
