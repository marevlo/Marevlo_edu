## Part 1 · Defining a Class and Creating Instances

A class is defined with the `class` keyword. The special method `__init__` is the **initialiser**: it runs automatically when a new instance is created and sets up that object's starting data. Its first parameter, by convention named `self`, refers to the instance being created, so `self.name = name` stores a value on that specific object.

Creating an instance looks like calling the class: `Dog("Rex")`. Python builds the object and passes it as `self`.

```python
class Dog:
    def __init__(self, name, age):
        # self refers to the new instance; these become its instance variables.
        self.name = name
        self.age = age

# Create two independent instances.
rex = Dog("Rex", 3)
bella = Dog("Bella", 5)

print(rex.name, "is", rex.age)
print(bella.name, "is", bella.age)
print("they are different objects:", rex is not bella)
```

Each instance carries its own data. Changing one does not affect the other, because their instance variables are stored separately on each object.

```python
rex.age = 4              # update only this instance
print("rex:", rex.age)   # 4
print("bella:", bella.age)   # 5, unchanged

# Attributes are stored in the instance's __dict__.
print("rex's stored data:", rex.__dict__)
```

## Part 2 · Instance Methods

A method is a function defined inside a class. An **instance method** takes `self` as its first parameter and can read and modify that instance's data. You call it on an instance with the dot syntax, and Python passes the instance as `self` automatically.

```python
class Counter:
    def __init__(self, start=0):
        self.value = start

    def increment(self, by=1):
        """Increase the counter and return the new value."""
        self.value += by      # modifies this instance's data
        return self.value

    def reset(self):
        self.value = 0

c = Counter()
print(c.increment())     # 1
print(c.increment(5))    # 6
c.reset()
print("after reset:", c.value)

# c.increment() is shorthand for Counter.increment(c): the instance becomes self.
```

## Part 3 · Class Variables Versus Instance Variables

An **instance variable** belongs to a single object and is set on `self`. A **class variable** is defined directly in the class body and is **shared by all instances**. Use class variables for data common to every instance, such as a default, a constant, or a shared counter.

```python
class Employee:
    company = "Acme Corp"      # class variable: shared by all employees
    count = 0                  # shared counter

    def __init__(self, name):
        self.name = name       # instance variable: unique per employee
        Employee.count += 1    # update the shared counter on the class

a = Employee("Asha")
b = Employee("Ben")

print(a.name, "works at", a.company)   # company comes from the class
print(b.name, "works at", b.company)
print("total employees:", Employee.count)

# Changing the class variable on the class affects everyone.
Employee.company = "Globex"
print("after rename:", a.company, b.company)
```

> **Caution — shared mutable class variables.** A mutable class variable such as a list is shared by every instance, which is rarely intended. Keep per-instance data on `self`, created inside `__init__`. The cell below shows the trap and the fix.

```python
class Buggy:
    items = []                 # shared by ALL instances; usually a mistake
    def add(self, x):
        self.items.append(x)

p, q = Buggy(), Buggy()
p.add("a")
print("q sees p's item:", q.items)   # ['a'] -- shared, surprising

class Fixed:
    def __init__(self):
        self.items = []        # a fresh list per instance
    def add(self, x):
        self.items.append(x)

r, s = Fixed(), Fixed()
r.add("a")
print("s is independent:", s.items)  # [] -- correct
```

## Part 4 · Class Methods and Static Methods

Beyond instance methods, classes support two other kinds, marked with decorators:

- A **class method** (`@classmethod`) receives the class itself as its first parameter, named `cls`. It is commonly used for alternative constructors that build an instance in a particular way.
- A **static method** (`@staticmethod`) receives neither `self` nor `cls`. It is a plain function grouped with the class because it is related to it logically.

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @classmethod
    def from_fahrenheit(cls, f):
        """An alternative constructor that converts before building the object."""
        return cls((f - 32) * 5 / 9)    # cls(...) creates a Temperature instance

    @staticmethod
    def is_freezing(celsius):
        """A related helper that needs no instance or class data."""
        return celsius <= 0

t1 = Temperature(25)                    # the usual constructor
t2 = Temperature.from_fahrenheit(212)   # the alternative constructor
print("t1:", t1.celsius, "| t2:", round(t2.celsius, 1))
print("is 25 freezing?", Temperature.is_freezing(25))
print("is -5 freezing?", Temperature.is_freezing(-5))
```

A quick guide to choosing: use an **instance method** when you need the object's data (`self`); a **class method** when you need the class but not a specific instance, such as an alternative constructor (`cls`); and a **static method** when you need neither but the function belongs with the class for organisation.

## Part 5 · Representing and Inspecting Objects

By default, printing an object shows an unhelpful memory address. Defining `__repr__` gives a readable, informative representation, which is invaluable for debugging. This is your first magic method; Module 3.3 covers many more.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        # A clear representation; ideally one that shows how to recreate the object.
        return f"Point(x={self.x}, y={self.y})"

p = Point(3, 4)
print(p)            # uses __repr__
print([p, Point(0, 0)])   # __repr__ is used inside containers too

# Inspection helpers:
print("type:", type(p).__name__)
print("is a Point:", isinstance(p, Point))
print("has attribute x:", hasattr(p, "x"))
```

### Example 1 — A simple class with data and a method (Easy)

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

r = Rectangle(4, 5)
print("area:", r.area())
# Experiment: create another rectangle and compare areas.
```

### Example 2 — Tracking state across method calls (Easy)

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("insufficient funds")

acc = BankAccount(100)
acc.deposit(50)
acc.withdraw(30)
print("balance:", acc.balance)
```

### Example 3 — A shared counter with a class variable (Medium)

```python
class Widget:
    total_created = 0           # shared across all widgets

    def __init__(self, label):
        self.label = label
        self.id = Widget.total_created     # assign before incrementing
        Widget.total_created += 1

w1 = Widget("button")
w2 = Widget("slider")
w3 = Widget("toggle")
print("ids:", w1.id, w2.id, w3.id)
print("total created:", Widget.total_created)
```

### Example 4 — An alternative constructor with a class method (Medium)

```python
class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    @classmethod
    def from_string(cls, text):
        """Build a Date from 'YYYY-MM-DD'."""
        year, month, day = (int(part) for part in text.split("-"))
        return cls(year, month, day)

    def __repr__(self):
        return f"Date({self.year}, {self.month}, {self.day})"

d1 = Date(2025, 5, 30)
d2 = Date.from_string("2025-12-25")
print(d1, "and", d2)
```

### Example 5 — Combining all three method kinds (Difficult)

```python
class Circle:
    pi = 3.141592653589793       # class variable used by instances

    def __init__(self, radius):
        self.radius = radius

    def area(self):                          # instance method: uses self
        return Circle.pi * self.radius ** 2

    @classmethod
    def unit(cls):                           # class method: alternative constructor
        return cls(1)

    @staticmethod
    def is_valid_radius(r):                  # static method: a related check
        return r > 0

c = Circle(2)
print("area:", round(c.area(), 2))
print("unit circle area:", round(Circle.unit().area(), 2))
print("valid radius -3?", Circle.is_valid_radius(-3))
```

### Example 6 — A small stock-keeping class with validation and representation (Difficult)

```python
class Product:
    catalogue = {}               # class-level registry of all products by code

    def __init__(self, code, name, price):
        self.code = code
        self.name = name
        self.price = price
        Product.catalogue[code] = self       # register this instance

    def apply_discount(self, percent):
        self.price *= (1 - percent / 100)
        return self.price

    @classmethod
    def find(cls, code):
        return cls.catalogue.get(code)

    def __repr__(self):
        return f"Product({self.code!r}, {self.name!r}, {self.price:.2f})"

Product("A1", "Notebook", 50)
Product("B2", "Pen", 10)

found = Product.find("A1")
print("found:", found)
found.apply_discount(20)
print("after discount:", found)
print("whole catalogue:", Product.catalogue)
```

**Solution 4**

```python
class Person:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    @classmethod
    def from_full_name(cls, full_name):
        first, last = full_name.split(" ", 1)
        return cls(first, last)

p = Person.from_full_name("Asha Patel")
print(p.first, p.last)
```

**Solution 5**

```python
class Playlist:
    def __init__(self):
        self.songs = []                  # fresh list per playlist

    def add(self, song):
        self.songs.append(song)

    def total(self):
        return len(self.songs)

    @staticmethod
    def is_valid_title(title):
        return bool(title.strip())

p = Playlist()
p.add("Song A")
p.add("Song B")
print("total:", p.total())
print("valid title '  '?", Playlist.is_valid_title("  "))
print("valid title 'Mix'?", Playlist.is_valid_title("Mix"))
```

## Summary and Key Points

- A class is a blueprint; an instance is an object built from it. `__init__` initialises a new instance, and `self` refers to that instance.
- Instance variables live on `self` and are unique per object; class variables are defined in the class body and shared by all instances. Avoid shared mutable class variables; create per-instance state in `__init__`.
- Instance methods take `self` and operate on object data. Class methods take `cls` (often alternative constructors); static methods take neither and are organisational helpers.
- Defining `__repr__` gives objects a clear, useful representation for debugging and display.
- Choose the method kind by what it needs: the instance, the class, or neither.
