## Part 1 · NumPy Arrays and Vectorisation

**NumPy** provides the `ndarray`, a fixed-type, multi-dimensional array stored compactly in memory (recall `array.array` from Module 7.2; NumPy generalises this and adds fast operations). Its key idea is **vectorisation**: operations apply to whole arrays at once, in optimised C, instead of element-by-element Python loops. This is both far faster and more concise.

```python
import numpy as np

# Create arrays from lists, or with helpers.
a = np.array([1, 2, 3, 4])
b = np.arange(0, 8, 2)            # [0, 2, 4, 6], like range but an array
zeros = np.zeros(3)
print("a:", a)
print("b:", b)

# Vectorised arithmetic: no loop needed; applies to every element at once.
print("a + 10:", a + 10)
print("a * b:", a * b)            # element-wise multiplication
print("a squared:", a ** 2)

# Reductions operate over the whole array.
print("sum:", a.sum(), "| mean:", a.mean(), "| max:", a.max())
```

```python
import numpy as np

# Arrays can be multi-dimensional. Shape describes the dimensions.
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("shape:", matrix.shape, "(2 rows, 3 columns)")
print("sum of all:", matrix.sum())
print("sum down columns (axis 0):", matrix.sum(axis=0))
print("sum across rows (axis 1):", matrix.sum(axis=1))

# Boolean masks select elements that satisfy a condition (vectorised filtering).
data = np.array([5, 12, 3, 18, 7, 20])
print("values > 10:", data[data > 10])
```

## Part 2 · Pandas DataFrames

**Pandas** builds on NumPy to handle **tabular** data: rows and named columns, like a spreadsheet or a database table. The central object is the **DataFrame**, and a single column is a **Series**. You can build one from a dictionary of columns, and inspect it with attributes such as `.shape`, `.columns`, `.dtypes`, and methods like `.head()` and `.describe()`.

```python
import pandas as pd

# A DataFrame from a dictionary: keys become column names.
df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dev", "Ela"],
    "city": ["Pune", "Delhi", "Pune", "Delhi", "Pune"],
    "age": [30, 25, 35, 28, 42],
    "salary": [70000, 50000, 90000, 60000, 110000],
})

print("shape:", df.shape)               # (rows, columns)
print("columns:", list(df.columns))
print()
print(df)
```

```python
import pandas as pd
df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dev", "Ela"],
    "age": [30, 25, 35, 28, 42],
    "salary": [70000, 50000, 90000, 60000, 110000],
})

# describe() gives quick summary statistics for numeric columns.
print(df.describe())
print()
# Access a single column (a Series) and compute on it.
print("mean age:", df["age"].mean())
print("max salary:", df["salary"].max())
```

## Part 3 · Selecting and Filtering

Pandas offers expressive selection. Pick columns by name, rows by position with `.iloc` or by label with `.loc`, and filter rows with a **boolean condition**, the tabular analogue of NumPy's masks. Filtering is how you answer questions like "which rows meet this criterion?"

```python
import pandas as pd
df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dev", "Ela"],
    "city": ["Pune", "Delhi", "Pune", "Delhi", "Pune"],
    "age": [30, 25, 35, 28, 42],
    "salary": [70000, 50000, 90000, 60000, 110000],
})

# Select specific columns.
print("names and ages:")
print(df[["name", "age"]])
print()

# Filter rows with a boolean condition.
older = df[df["age"] > 30]
print("people over 30:")
print(older[["name", "age"]])
print()

# Combine conditions with & (and) and | (or); each condition in parentheses.
pune_high = df[(df["city"] == "Pune") & (df["salary"] > 80000)]
print("high earners in Pune:")
print(pune_high[["name", "salary"]])
```

## Part 4 · Grouping and Aggregating

A core data task is **grouping** rows by a category and computing a summary per group, the same idea as SQL's `GROUP BY` (Module 8.1) and `itertools.groupby` (Module 4.3), but far more convenient. `df.groupby("column")` followed by an aggregation gives one result per group.

```python
import pandas as pd
df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dev", "Ela"],
    "city": ["Pune", "Delhi", "Pune", "Delhi", "Pune"],
    "age": [30, 25, 35, 28, 42],
    "salary": [70000, 50000, 90000, 60000, 110000],
})

# Average salary per city.
print("mean salary by city:")
print(df.groupby("city")["salary"].mean())
print()

# Several aggregations at once with .agg.
summary = df.groupby("city").agg(
    headcount=("name", "count"),
    avg_age=("age", "mean"),
    total_salary=("salary", "sum"),
)
print("summary by city:")
print(summary)
```

## Part 5 · Plotting with matplotlib

**matplotlib** is the foundational plotting library. The usual pattern creates a figure and axes, calls a plotting method, labels it, and shows or saves it. In a notebook plots normally appear inline; here we save each to an image file and confirm it was written, so the examples run without a display. Pandas also has a convenient `.plot()` that uses matplotlib under the hood.

```python
import matplotlib
matplotlib.use("Agg")                 # a non-interactive backend that renders to files
import matplotlib.pyplot as plt
import numpy as np
import os

# A simple line plot.
x = np.linspace(0, 10, 100)           # 100 points from 0 to 10
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y, label="sin(x)")
ax.set_title("A sine curve")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()
fig.tight_layout()

fig.savefig("sine.png", dpi=80)
plt.close(fig)
print("saved sine.png:", os.path.getsize("sine.png"), "bytes")
print("in a notebook with an inline backend, this figure would display directly")
```

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import pandas as pd
import os

df = pd.DataFrame({
    "city": ["Pune", "Delhi", "Mumbai"],
    "sales": [120, 95, 150],
})

# A bar chart from a DataFrame, using matplotlib directly.
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(df["city"], df["sales"], color="steelblue")
ax.set_title("Sales by city")
ax.set_ylabel("sales")
fig.tight_layout()
fig.savefig("sales.png", dpi=80)
plt.close(fig)
print("saved sales.png:", os.path.getsize("sales.png"), "bytes")

# Pandas can plot directly too: df.plot(kind="bar") would produce a similar chart.
print("equivalently, df.plot(kind='bar', x='city', y='sales') uses matplotlib internally")
```

### Example 2 — A basic DataFrame (Easy)

### Example 3 — Filtering rows (Medium)

```python
import pandas as pd
df = pd.DataFrame({
    "product": ["a", "b", "c", "d"],
    "stock": [0, 12, 3, 25],
})
in_stock = df[df["stock"] > 0]
print("products in stock:")
print(in_stock)
print("low stock (< 5):", list(df[df["stock"] < 5]["product"]))
```

### Example 4 — Grouping and summing (Medium)

```python
import pandas as pd
df = pd.DataFrame({
    "region": ["N", "S", "N", "S", "N"],
    "sales": [10, 20, 15, 25, 30],
})
print("total sales by region:")
print(df.groupby("region")["sales"].sum())
```

### Example 5 — A small analysis pipeline (Difficult)

```python
import pandas as pd

# Raw data: orders with a product, quantity, and unit price.
df = pd.DataFrame({
    "product": ["pen", "book", "pen", "lamp", "book", "pen"],
    "qty": [3, 1, 5, 2, 4, 2],
    "unit_price": [1.5, 4.0, 1.5, 12.0, 4.0, 1.5],
})

# Derive a revenue column, then summarise per product.
df["revenue"] = df["qty"] * df["unit_price"]
report = df.groupby("product").agg(
    units=("qty", "sum"),
    revenue=("revenue", "sum"),
).sort_values("revenue", ascending=False)
print(report)
print()
print("best-selling by revenue:", report.index[0])
```

### Example 6 — Plotting grouped results (Difficult)

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import pandas as pd
import os

df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr"],
    "revenue": [200, 350, 300, 500],
})

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(df["month"], df["revenue"], marker="o")
ax.set_title("Monthly revenue")
ax.set_ylabel("revenue")
ax.grid(True, alpha=0.3)
fig.tight_layout()
fig.savefig("revenue.png", dpi=80)
plt.close(fig)
print("saved revenue.png:", os.path.getsize("revenue.png"), "bytes")
print("growth from Jan to Apr:", df["revenue"].iloc[-1] - df["revenue"].iloc[0])
```

## Summary and Key Points

- NumPy's `ndarray` stores fixed-type data compactly and supports vectorisation: operations apply to whole arrays in fast C, replacing element-wise Python loops; reductions (`sum`, `mean`) and boolean masks operate across arrays and support multiple dimensions via `axis`.
- Pandas builds on NumPy for tabular data: the DataFrame holds named columns (each a Series), built easily from a dict, with `.shape`, `.describe()`, and `.head()` for inspection.
- Select columns by name, rows by `.iloc`/`.loc`, and filter rows with boolean conditions combined using `&` and `|` (each parenthesised).
- `groupby` plus an aggregation (or `.agg` for several) summarises data per category, the tabular equivalent of SQL `GROUP BY`.
- matplotlib plots via a figure-and-axes pattern (`plot`, `bar`, labels, then show or save); Pandas `.plot()` wraps it for quick charts. This stack is the launchpad for data analysis.
