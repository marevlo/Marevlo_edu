```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
# Create synthetic customer data for analysis
# This dataset represents different customer segments with their spending patterns
np.random.seed(42)
n_customers = 150

customer_data = pd.DataFrame({
    'customer_id': range(1, n_customers + 1),
    'annual_income': np.random.normal(50000, 15000, n_customers),
    'spending_score': np.random.normal(50, 20, n_customers),
    'age': np.random.normal(35, 10, n_customers),
    'segment': np.random.choice(['Low Value', 'Medium Value', 'High Value'], n_customers,
                                p=[0.3, 0.5, 0.2])
})

# Map segments to numeric values for coloring
segment_mapping = {'Low Value': 0, 'Medium Value': 1, 'High Value': 2}
customer_data['segment_code'] = customer_data['segment'].map(segment_mapping)

customer_data.sample(5)
```

```python
# The segment_code column already contains numeric values for coloring
# This allows us to use different colors for each customer segment
customer_data.sample(5)
```

```python
# Create a colored scatter plot showing customer segments
# Each color represents a different customer segment based on their value to the business
plt.scatter(customer_data['annual_income'], customer_data['spending_score'],
           c=customer_data['segment_code'], cmap='viridis', alpha=0.7, s=50)
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation Analysis')
plt.colorbar(label='Customer Segment\n(0=Low, 1=Medium, 2=High)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# The cmap parameter controls the color map, and alpha controls transparency
# s parameter controls the size of points, c parameter colors by the segment_code
```

### Plot size

```python
# Controlling plot size with figsize parameter
# figsize=(width, height) in inches - larger figures are better for presentations
plt.figure(figsize=(12, 8))

plt.scatter(customer_data['annual_income'], customer_data['spending_score'],
           c=customer_data['segment_code'], cmap='plasma', alpha=0.8, s=60,
           edgecolors='black', linewidth=0.5)
plt.xlabel('Annual Income ($)', fontsize=12)
plt.ylabel('Spending Score (1-100)', fontsize=12)
plt.title('Customer Segmentation Analysis - Large Format', fontsize=14, fontweight='bold')
plt.colorbar(label='Customer Segment\n(0=Low, 1=Medium, 2=High)')
plt.grid(True, alpha=0.3)
plt.tight_layout()  # Prevents labels from being cut off
plt.show()
```

### Annotations

```python
# Create employee performance dataset for annotations demonstration
np.random.seed(42)
n_employees = 50

employees = pd.DataFrame({
    'employee_id': range(1, n_employees + 1),
    'name': [f'Employee_{i}' for i in range(1, n_employees + 1)],
    'productivity_score': np.random.normal(75, 15, n_employees),
    'efficiency_rating': np.random.normal(80, 12, n_employees),
    'performance_points': np.random.normal(150, 40, n_employees)
})

# Clip values to reasonable ranges
employees['productivity_score'] = np.clip(employees['productivity_score'], 30, 100)
employees['efficiency_rating'] = np.clip(employees['efficiency_rating'], 40, 100)
employees['performance_points'] = np.clip(employees['performance_points'], 50, 250)
```

```python
employees.shape
```

```python
# Select a random sample of employees for visualization
# This helps avoid cluttering when dealing with large datasets
sample_df = employees.sample(20, random_state=42)
```

```python
sample_df
```

```python
# Create scatter plot with text annotations
# Annotations help identify specific data points of interest
plt.figure(figsize=(16, 10))

# Use performance_points for bubble size to show overall performance
plt.scatter(sample_df['productivity_score'], sample_df['efficiency_rating'],
           s=sample_df['performance_points'], alpha=0.7, c='skyblue',
           edgecolors='navy', linewidth=1)

plt.title('Employee Performance Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Productivity Score', fontsize=12)
plt.ylabel('Efficiency Rating', fontsize=12)
plt.grid(True, alpha=0.3)

# Add employee names as annotations
for i in range(sample_df.shape[0]):
    plt.text(sample_df['productivity_score'].values[i],
             sample_df['efficiency_rating'].values[i],
             sample_df['name'].values[i],
             fontsize=8, ha='center', va='bottom',
             bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

plt.show()
```

```python
# Simple text annotations on a basic scatter plot
# plt.text() adds text at specific coordinates
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.scatter(x, y, s=100, c='red', alpha=0.7)
plt.text(1, 5, 'Point 1', fontsize=12, color='blue', ha='left', va='bottom')
plt.text(2, 6, 'Point 2', fontsize=12, color='green', ha='center', va='top')
plt.text(3, 7, 'Point 3', fontsize=12, color='purple', ha='right', va='bottom',
         bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5))
plt.text(4, 8, 'Point 4', fontsize=12, color='brown', fontweight='bold')

plt.title('Basic Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, alpha=0.3)
plt.show()
```

### 3D Surface Plots

```python
# Create coordinate grids for 3D surface plotting
# X and Y represent independent variables, Z is the dependent variable
x = np.linspace(-5, 5, 50)  # 50 points from -5 to 5
y = np.linspace(-5, 5, 50)  # 50 points from -5 to 5
```

```python
# Create 2D coordinate grids using meshgrid
# This creates matrices where XX contains x-coordinates and YY contains y-coordinates
XX, YY = np.meshgrid(x, y)
```

```python
# Calculate Z values for each (X,Y) point - this creates a 3D surface
# Using a quadratic function: Z = X² + Y² (creates a parabolic bowl shape)
z = XX**2 + YY**2
z.shape  # Should be (50, 50) - same dimensions as the meshgrid
```

```python
# Plot 3D surface using the quadratic function Z = X² + Y²
# This creates a parabolic surface that opens upwards
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')

surface = ax.plot_surface(XX, YY, z, cmap='viridis', alpha=0.8)
fig.colorbar(surface, shrink=0.6)
ax.set_title('3D Surface Plot: Z = X² + Y²', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=10)
ax.set_ylabel('Y values', fontsize=10)
ax.set_zlabel('Z values', fontsize=10)

plt.show()
```

```python
# Plot a more complex surface using trigonometric functions
# Z = sin(X) + cos(Y) creates a wavy surface
z_trig = np.sin(XX) + np.cos(YY)

fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')

surface = ax.plot_surface(XX, YY, z_trig, cmap='plasma', alpha=0.8)
fig.colorbar(surface, shrink=0.6)
ax.set_title('3D Surface Plot: Z = sin(X) + cos(Y)', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=10)
ax.set_ylabel('Y values', fontsize=10)
ax.set_zlabel('Z values', fontsize=10)

plt.show()
```

```python
# Another complex surface with exponential decay
# Z = sin(X) + log(X) - this will have issues where X ≤ 0 due to log
try:
    z_log = np.sin(XX) + np.log(XX + 6)  # Add 6 to avoid log(0) or negative values

    fig = plt.figure(figsize=(12, 8))
    ax = plt.subplot(projection='3d')

    surface = ax.plot_surface(XX, YY, z_log, cmap='coolwarm', alpha=0.8)
    fig.colorbar(surface, shrink=0.6)
    ax.set_title('3D Surface Plot: Z = sin(X) + log(X+6)', fontsize=14, fontweight='bold')
    ax.set_xlabel('X values', fontsize=10)
    ax.set_ylabel('Y values', fontsize=10)
    ax.set_zlabel('Z values', fontsize=10)

    plt.show()
except Exception as e:
    print(f"Error in surface plot: {e}")
    print("This can happen with mathematical functions that have domain restrictions")
```

```python
# Alternative surface plot with different viewing angle
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')

surface = ax.plot_surface(XX, YY, z_trig, cmap='viridis', alpha=0.8)
fig.colorbar(surface, shrink=0.6)

# Change the viewing angle
ax.view_init(elev=30, azim=45)  # elev=30° above horizontal, azim=45° rotation
ax.set_title('3D Surface Plot: Different Viewing Angle', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=10)
ax.set_ylabel('Y values', fontsize=10)
ax.set_zlabel('Z values', fontsize=10)

plt.show()
```

### Contour Plots

```python
# Contour Plot - 2D representation of 3D surfaces
# Shows level curves of the Z function on the X-Y plane
fig = plt.figure(figsize=(10, 8))
ax = plt.subplot()

# Create contour lines
contours = ax.contour(XX, YY, z_trig, cmap='viridis', levels=15)
fig.colorbar(contours, shrink=0.8)
ax.set_title('Contour Plot: Level Curves of Z = sin(X) + cos(Y)', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.grid(True, alpha=0.3)

plt.show()
```

```python
# Filled Contour Plot - adds color between contour lines
# This creates a heatmap-like visualization of the surface
fig = plt.figure(figsize=(10, 8))
ax = plt.subplot()

# Create filled contours
filled_contours = ax.contourf(XX, YY, z_trig, cmap='viridis', levels=20, alpha=0.8)
fig.colorbar(filled_contours, shrink=0.8)
ax.set_title('Filled Contour Plot: Z = sin(X) + cos(Y)', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.grid(True, alpha=0.3)

plt.show()
```

```python
# Filled contour with different function
z_wave = np.sin(XX) * np.cos(YY)

fig = plt.figure(figsize=(10, 8))
ax = plt.subplot()

filled_contours = ax.contourf(XX, YY, z_wave, cmap='RdYlBu', levels=15, alpha=0.8)
fig.colorbar(filled_contours, shrink=0.8)
ax.set_title('Filled Contour Plot: Z = sin(X) × cos(Y)', fontsize=14, fontweight='bold')
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.grid(True, alpha=0.3)

plt.show()
```

```python
# Create synthetic sales data for heatmap visualization
# This represents sales performance across different regions and quarters
np.random.seed(42)

regions = ['North', 'South', 'East', 'West', 'Central']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
products = ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard']

# Create sales data (rows = regions, columns = quarters)
sales_data = np.random.randint(1000, 5000, size=(len(regions), len(quarters)))

# Create DataFrame for easier manipulation
sales_df = pd.DataFrame(sales_data, index=regions, columns=quarters)
print("Regional Sales Data (in $1000s):")
sales_df
```
