## What is Pandas Series?

A **Pandas Series** is a 1-dimensional array with **labeled indices**. Think of it as a single column in a spreadsheet with:

- **Values**: The actual data (numbers, strings, dates, etc.)
- **Index**: Labels for each value (can be integers, strings, dates, etc.)

Series are the foundation of DataFrames (which are collections of Series).

## Part 1: Getting Started with Series

### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print(f"Pandas Version: {pd.__version__}")
```

### Step 2: Load Netflix Data

```python
# Create sample Netflix-like dataset locally (no external URL needed)
import random

random.seed(42)
np.random.seed(42)

# Generate realistic sample data
titles = [
    'Breaking Bad', 'The Office', 'Stranger Things', 'Friends', 'The Crown',
    'Narcos', 'The Good Place', 'Chernobyl', 'Mindhunter', 'Dark',
    'The Witcher', 'Ozark', 'Peaky Blinders', 'Handmaid\'s Tale', 'Bodyguard',
    'Unbreakable Kimmy Schmidt', 'Schitt\'s Creek', 'Brooklyn Nine-Nine', 'The Marvelous Mrs. Maisel', 'Fleabag'
]

netflix = pd.DataFrame({
    'title': titles,
    'type': [random.choice(['Movie', 'TV Show']) for _ in titles],
    'release_year': np.random.randint(2015, 2024, len(titles)),
    'rating': [random.choice(['TV-MA', 'TV-14', 'PG-13', 'R', 'PG', 'G', 'NC-17']) for _ in titles],
    'duration': [f"{random.randint(40, 180)} min" if random.choice([True, False]) else f"{random.randint(1, 5)} Seasons" for _ in titles],
    'country': [random.choice(['United States', 'United Kingdom', 'South Korea', 'Canada', 'Germany']) for _ in titles]
})

print(f"Dataset shape: {netflix.shape}")
print(f"\nFirst few rows:")
print(netflix.head())
```

```python
# Check data types and missing values
print(netflix.info())
```

## Part 2: Creating Series

### 2A: Series from a List

```python
# Simple Series from list (auto-indexed with 0, 1, 2, ...)
imdb_scores = [8.2, 7.1, 8.9, 6.5, 9.1]
scores_series = pd.Series(imdb_scores)

print("Series from list (default index):")
print(scores_series)
print(f"\nType: {type(scores_series)}")
```

```python
# Series with custom index (more meaningful!)
show_titles = ['Breaking Bad', 'The Office', 'Stranger Things', 'Friends', 'The Crown']
imdb_scores = [9.5, 9.0, 8.4, 8.9, 8.6]

shows = pd.Series(imdb_scores, index=show_titles, name='IMDB Ratings')

print("Series with custom index:")
print(shows)
print(f"\nName of Series: {shows.name}")
```

### 2B: Series from a Dictionary

```python
# Dictionary keys become index, values become Series values
netflix_by_country = {
    'United States': 2555,
    'India': 972,
    'United Kingdom': 627,
    'Canada': 403,
    'Spain': 365
}

content_count = pd.Series(netflix_by_country, name='Content Count by Country')
print(content_count)
```

### 2C: Extract Series from DataFrame Column

```python
# Extract a single column as Series
titles_series = netflix['title']

print(f"Type of netflix['title']: {type(titles_series)}")
print(f"Shape: {titles_series.shape}")
print(f"\nFirst 5 titles:")
print(titles_series.head())
```

## Part 3: Understanding Series Attributes

```python
# Use our Netflix shows Series
print("\n=== SERIES ATTRIBUTES ===")
print(f"Shows Series:\n{shows}")
print(f"\n{'─'*40}")
print(f"Name:        {shows.name}")
print(f"Index:       {shows.index.tolist()}")
print(f"Values:      {shows.values}")
print(f"Data Type:   {shows.dtype}")
print(f"Size:        {shows.size}")
print(f"Shape:       {shows.shape}")
print(f"Memory use:  {shows.memory_usage()} bytes")
```

## Part 4: Indexing and Accessing Data

### 4A: Integer-based Indexing (Position)

```python
# Access by position using iloc
print("First show (position 0):")
print(shows.iloc[0])

print("\nLast show (position -1):")
print(shows.iloc[-1])

print("\nPositions 1-3:")
print(shows.iloc[1:3])  # Slicing: excludes the end index
```

### 4B: Label-based Indexing

```python
# Access by label using loc
print("Rating for Breaking Bad:")
print(shows.loc['Breaking Bad'])

print("\nMultiple shows:")
print(shows.loc[['Breaking Bad', 'Stranger Things']])

print("\nAll shows with rating > 8.5:")
print(shows[shows > 8.5])
```

### 4C: Conditional Filtering

```python
# Filter Netflix DataFrame
rating_series = netflix['rating'].dropna()

# Shows rated 'TV-MA' (Mature Audiences)
print("Distribution of ratings:")
print(rating_series.value_counts().head(10))

# Boolean indexing
tv_shows = netflix[netflix['type'] == 'TV Show']
print(f"\nNumber of TV Shows: {len(tv_shows)}")
print(f"Number of Movies: {len(netflix[netflix['type'] == 'Movie'])}")
```

> 💡 **Best Practice**: Use `.loc[]` for label-based access and `.iloc[]` for position-based access. This makes your code more readable and less error-prone.

## Part 5: Methods for Analysis & Transformation

### 5A: Descriptive Statistics

```python
# Let's create a numeric Series from Netflix
# Extract release year from 'release_year' column
years = netflix['release_year'].dropna()

print("Descriptive statistics:")
print(years.describe())

print(f"\nMean: {years.mean():.2f}")
print(f"Median: {years.median():.2f}")
print(f"Std Dev: {years.std():.2f}")
print(f"Min: {years.min():.0f}")
print(f"Max: {years.max():.0f}")
```

### 5B: Unique & Value Counts

```python
# Unique values in a Series
types = netflix['type'].dropna()

print(f"Unique types: {types.unique()}")
print(f"Number of unique types: {types.nunique()}")

print("\nValue counts (sorted by frequency):")
print(types.value_counts())

# Percentage distribution
print("\nPercentage distribution:")
print((types.value_counts() / len(types) * 100).round(1))
```

### 5C: The Apply Method - Transform Data

```python
# Example 1: Apply a custom function
ratings = pd.Series([8.5, 7.2, 9.1, 6.8, 8.9])

# Categorize ratings
def categorize_rating(rating):
    if rating >= 8.5:
        return 'Excellent'
    elif rating >= 7.0:
        return 'Good'
    else:
        return 'Average'

categories = ratings.apply(categorize_rating)
print("Original ratings:")
print(ratings.values)
print("\nCategorized:")
print(categories.values)
```

```python
# Example 2: Apply lambda (anonymous function)
years = netflix['release_year'].dropna()

# Calculate content age (2024 - release_year)
content_age = years.apply(lambda year: 2024 - int(year))

print("Content age distribution:")
print(content_age.describe())
print(f"\nNewest content: {content_age.min():.0f} years old")
print(f"Oldest content: {content_age.max():.0f} years old")
```

### 5D: The isin Method - Filter by Multiple Values

```python
# isin: Check if Series values are in a list
types = netflix['type']

# Filter for specific types
is_tv_or_movie = types.isin(['TV Show', 'Movie'])
print(f"Count of TV Show or Movie: {is_tv_or_movie.sum()}")

# Find shows from specific countries (if data available)
countries = netflix['country'].dropna()
usa_uk_india = countries.str.contains('United States|United Kingdom|India', na=False, regex=True)
print(f"\nContent from USA, UK, or India: {usa_uk_india.sum()}")
```

### 5E: String Methods (for text Series)

```python
# String operations on text Series
titles = netflix['title'].dropna().head(10)

print("Original titles:")
print(titles.values)

print("\nTitles (uppercase):")
print(titles.str.upper().values)

print("\nTitle length:")
print(titles.str.len().values)

# Check if title contains 'The'
has_the = titles.str.contains('The', case=False, na=False)
print(f"\nTitles containing 'The': {has_the.sum()}")
```

> ⚠️ **Common Pitfall**: String methods on a Series use `.str` accessor. Forgetting this will cause an error. For example, `titles.upper()` ✗ → `titles.str.upper()` ✓

## Part 6: Handling Missing Data

```python
# Check for missing values
print("Missing values per column:")
print(netflix.isnull().sum())

# Check percentage of missing data
print("\nPercentage missing:")
print((netflix.isnull().sum() / len(netflix) * 100).round(2))
```

```python
# Method 1: dropna - Remove missing values
country = netflix['country']
print(f"Original size: {len(country)}")
print(f"Missing values: {country.isnull().sum()}")

country_clean = country.dropna()
print(f"After dropna: {len(country_clean)}")
```

```python
# Method 2: fillna - Replace missing values
duration = netflix['duration']
print(f"Sample of duration with missing:")
print(duration.head(10))

# Fill with a default value
duration_filled = duration.fillna('Unknown')
print(f"\nAfter fillna('Unknown'):")
print(duration_filled.head(10))
```

```python
# Fill with forward-fill or backward-fill
numeric_series = pd.Series([1, 2, np.nan, 4, np.nan, 6])

print("Original:")
print(numeric_series.values)

print("\nForward-fill (propagate previous value):")
print(numeric_series.ffill().values)

print("\nBackward-fill (propagate next value):")
print(numeric_series.bfill().values)
```

## Part 7: Vectorized Operations & Performance

```python
# Vectorized operations are MUCH faster than loops
years = netflix['release_year'].dropna()

# VECTORIZED (Fast!) ✓
ages_vectorized = 2024 - years
print(f"Vectorized result (first 5): {ages_vectorized.head().values}")

print("\nVectorized operations are ~100x faster than loops!")
```

```python
# Arithmetic operations on Series
scores = pd.Series([7.5, 8.2, 6.9, 9.1], index=['Show A', 'Show B', 'Show C', 'Show D'])

print("Original scores:")
print(scores)

print("\nScores scaled by 10:")
print(scores * 10)

print("\nScore deviations from mean:")
mean_score = scores.mean()
print(scores - mean_score)
```

> 🔍 **Real-World Use Case**: When working with datasets containing thousands or millions of rows, vectorized operations are essential. A loop-based approach could take minutes, while vectorization completes in milliseconds!

## Part 8: Method Chaining (Write Elegant Code)

```python
# WITHOUT chaining (harder to read):
country = netflix['country'].dropna()
country_upper = country.str.upper()
country_stripped = country_upper.str.strip()
result1 = country_stripped.value_counts().head(5)

# WITH chaining (clean & readable):
result2 = (netflix['country']
           .dropna()
           .str.upper()
           .str.strip()
           .value_counts()
           .head(5))

print("Top 5 countries (chained):")
print(result2)
```

## Part 9: Try It Yourself!

### Challenge 1: Create a Series of Netflix Release Years

Create a Series from the 'release_year' column and:
1. Find the mean, median, and standard deviation
2. Filter for years >= 2020
3. Count how many shows/movies were released in 2020

```python
# Challenge 1 Solution Steps:

# STEP 1: Create Series from 'release_year' column
release_years = netflix['release_year']
print(f"Release years range: {release_years.min()} to {release_years.max()}")

# STEP 2: Filter for years >= 2020
recent_releases = release_years[release_years >= 2020]
print(f"\nReleases from 2020 onward: {len(recent_releases)} items")
print(f"Percentage of total: {len(recent_releases) / len(release_years) * 100:.1f}%")

# STEP 3: Count and analyze
year_distribution = recent_releases.value_counts().sort_index()
print("\nBreakdown by year:")
print(year_distribution)

# BONUS: Combine with type to see content strategy
netflix_recent = netflix[netflix['release_year'] >= 2020]
print("\n~ Content type distribution in recent releases ~")
print(netflix_recent['type'].value_counts())
```

### Challenge 2: Analyze Netflix Ratings

Create a Series from the 'rating' column and:
1. Find the top 5 most common ratings
2. Calculate the percentage of content in each rating category
3. Create a new Series that flags 'explicit' content (ratings containing 'MA' or 'R')

```python
# Challenge 2 Solution Steps:

# STEP 1: Create a Series from 'rating' column
ratings = netflix['rating'].dropna()  # Remove any NaN values
print("Top 5 most common ratings:")
top_5_ratings = ratings.value_counts().head(5)
print(top_5_ratings)

# STEP 2: Calculate percentages
print("\n~ Rating distribution (percentage) ~")
rating_pct = (ratings.value_counts(normalize=True) * 100).round(1)
print(rating_pct.head(10))

# STEP 3: Flag explicit content (MA or R ratings)
explicit_flags = ratings.str.contains('MA|R', regex=True, na=False)
explicit_count = explicit_flags.sum()
print(f"\n~ Explicit Content Analysis ~")
print(f"Explicit content (MA or R): {explicit_count}")
print(f"Family-friendly: {(~explicit_flags).sum()}")
print(f"Explicit percentage: {explicit_count / len(ratings) * 100:.1f}%")

# BONUS: Create an explicit content indicator series
netflix['is_explicit'] = netflix['rating'].str.contains('MA|R', regex=True, na=False)
print(f"\nExplicit content ratio by type:")
print(netflix.groupby('type')['is_explicit'].mean().mul(100).round(1))
```
