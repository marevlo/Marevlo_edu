## What is a Pandas DataFrame?

A **DataFrame** is a 2-dimensional table with:

- **Rows**: Observations or records (indexed)
- **Columns**: Variables or features (labeled)
- **Values**: The actual data

Think of it as an Excel spreadsheet in Python! Each column is a Series, and a DataFrame is a collection of Series with the same index.

## Part 2: Creating DataFrames

### 2A: DataFrame from a Dictionary of Lists

```python
# Most common way: dictionary with column names as keys
shows_data = {
    'Title': ['Breaking Bad', 'The Office', 'Stranger Things', 'Friends', 'The Crown'],
    'Release Year': [2008, 2005, 2016, 1994, 2016],
    'IMDB Rating': [9.5, 9.0, 8.4, 8.9, 8.6],
    'Episodes': [62, 201, 42, 236, 50]
}

df_shows = pd.DataFrame(shows_data)
print("DataFrame from dictionary:")
print(df_shows)
print(f"\nType: {type(df_shows)}")
```

### 2B: DataFrame with Custom Index

```python
# Add meaningful index instead of default 0, 1, 2, ...
df_shows_indexed = pd.DataFrame(
    shows_data,
    index=['s1', 's2', 's3', 's4', 's5']
)

print("DataFrame with custom index:")
print(df_shows_indexed)
```

### 2C: DataFrame from Lists

```python
# List of lists (each row is a list)
data = [
    ['Alice', 25, 'Engineer', 120000],
    ['Bob', 30, 'Manager', 135000],
    ['Charlie', 23, 'Analyst', 95000]
]

columns = ['Name', 'Age', 'Position', 'Salary']
df_employees = pd.DataFrame(data, columns=columns)

print(df_employees)
```

### 2D: DataFrame from Combined Series

```python
# Create a DataFrame from multiple Series
sales = pd.Series([100, 200, 150, 250], name='Sales')
region = pd.Series(['North', 'South', 'East', 'West'], name='Region')
profit = pd.Series([20, 45, 25, 60], name='Profit')

# Concatenate Series into DataFrame
df_combined = pd.concat([sales, region, profit], axis=1)
print(df_combined)
```

## Part 8: Boolean Indexing & Filtering

### 8A: Single Condition

```python
# Filter Netflix for Movies only
movies = netflix[netflix['type'] == 'Movie']
print(f"Number of movies: {len(movies)}")
print(f"\nFirst few movies:")
print(movies[['title', 'type', 'release_year']].head())
```

```python
# Filter for recent content
recent = netflix[netflix['release_year'] >= 2020]
print(f"Content added in 2020 or later: {len(recent)}")
print(f"\nYears distribution:")
print(recent['release_year'].value_counts().sort_index())
```

### 8B: Multiple Conditions

```python
# AND condition (both must be true)
movies_recent = netflix[(netflix['type'] == 'Movie') & (netflix['release_year'] >= 2020)]
print(f"Movies released in 2020+: {len(movies_recent)}")

# OR condition (at least one must be true)
top_types = netflix[(netflix['type'] == 'Movie') | (netflix['type'] == 'TV Show')]
print(f"\nMovies or TV Shows: {len(top_types)}")
```

```python
# NOT condition
not_movies = netflix[~(netflix['type'] == 'Movie')]
print(f"Not movies: {len(not_movies)}")

# Using isin for multiple values
tv_rated = netflix[netflix['rating'].isin(['TV-MA', 'TV-14', 'TV-PG'])]
print(f"\nTV-rated content: {len(tv_rated)}")
```

> ⚠️ **Common Pitfall**: Use `&` (AND), `|` (OR), `~` (NOT) for boolean indexing, NOT `and`, `or`, `not`. These operators don't work with Series/DataFrames.

## Part 9: Working with Superstore Data (Advanced Exploration)

```python
# Preview Superstore data
print(f"Superstore shape: {superstore.shape}")
print(f"\nColumns: {superstore.columns.tolist()}")
print(f"\nFirst few rows:")
print(superstore.head())
```

```python
# Real-world question: What's the average sales by region?
if 'Region' in superstore.columns and 'Sales' in superstore.columns:
    sales_by_region = superstore.groupby('Region')['Sales'].mean()
    print("Average Sales by Region:")
    print(sales_by_region.sort_values(ascending=False))
```

```python
# What products are most profitable?
if 'Product Sub-Category' in superstore.columns and 'Profit' in superstore.columns:
    profit_by_product = superstore.groupby('Product Sub-Category')['Profit'].sum()
    print("\nTop 10 most profitable products:")
    print(profit_by_product.nlargest(10))
```

## Part 10: DataFrame Operations

### 10A: Adding Columns

```python
# Create a copy to avoid modifying original
netflix_copy = netflix.copy()

# Add a new column
netflix_copy['age_years'] = 2024 - netflix_copy['release_year']

print("New column 'age_years' added:")
print(netflix_copy[['title', 'release_year', 'age_years']].head())
```

```python
# Add column based on conditions
netflix_copy['is_new'] = netflix_copy['release_year'] >= 2020

print("\nNew 'is_new' column:")
print(netflix_copy[['title', 'release_year', 'is_new']].head(10))
```

### 10B: Dropping Columns & Rows

```python
# Drop columns
df_subset = netflix_copy.drop(columns=['age_years', 'is_new'])
print(f"Columns after drop: {len(df_subset.columns)} columns")
print(f"Original had: {len(netflix_copy.columns)} columns")

# Drop rows with missing values in a specific column
clean_ratings = netflix[netflix['rating'].notna()]
print(f"\nAfter removing rows with missing ratings:")
print(f"Original rows: {len(netflix)}")
print(f"Clean rows: {len(clean_ratings)}")
```

### 10C: Renaming Columns

```python
# Rename specific columns
netflix_renamed = netflix.rename(columns={
    'title': 'Title',
    'type': 'Content Type',
    'release_year': 'Year'
})

print("Renamed columns:")
print(netflix_renamed[['Title', 'Content Type', 'Year']].head())
```

## Part 11: Data Type Conversion

```python
# Check current data types
print("Current data types:")
print(netflix.dtypes)

# Convert to more efficient types
netflix_optimized = netflix.copy()
netflix_optimized['release_year'] = netflix_optimized['release_year'].astype('int16')
netflix_optimized['type'] = netflix_optimized['type'].astype('category')

print("\nMemory usage BEFORE optimization:")
print(f"{netflix.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print("\nMemory usage AFTER optimization:")
print(f"{netflix_optimized.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

> 🔍 **Real-World Use Case**: Optimizing data types can reduce memory usage by 50%+ on large datasets. This is crucial when working with limited RAM or deploying models to production!

## Part 12: Try It Yourself!

### Challenge 1: Netflix Analysis

Using the Netflix dataset:
1. How many TV Shows vs Movies?
2. What's the most common rating?
3. Find all content from 2023

```python
# Challenge 1: Netflix Analysis

print("NETFLIX DATAFRAME ANALYSIS")
print("=" * 60)

# Q1: How many TV Shows vs Movies?
print("\n1️⃣ Content Type Distribution:")
type_counts = netflix['type'].value_counts()
print(type_counts)
print(f"\nPercentage breakdown:")
print((netflix['type'].value_counts(normalize=True) * 100).round(1))

# Q2: What's the most common rating?
print("\n2️⃣ Most common content rating:")
common_rating = netflix['rating'].value_counts().head(1)
print(common_rating)
print(f"\nTop 5 ratings:")
print(netflix['rating'].value_counts().head(5))

# Q3: Find all content from 2023
print("\n3️⃣ Content from 2023:")
netflix_2023 = netflix[netflix['release_year'] == 2023]
print(f"Number of titles: {len(netflix_2023)}")
print(f"Type distribution:")
print(netflix_2023['type'].value_counts())

print("\n📊 Sample 2023 releases:")
print(netflix_2023[['title', 'type', 'rating']].head(10))
```

### Challenge 2: Superstore Analysis  

Using the Superstore dataset:
1. Create a new column 'Margin' = Profit / Sales * 100
2. Which region has the highest average profit?
3. Filter for profitable transactions (Profit > 0)

```python
# Challenge 2: Superstore Analysis

print("SUPERSTORE DATAFRAME ANALYSIS")
print("=" * 60)

# Q1: Create a new 'Margin' column
superstore['Margin%'] = (superstore['Profit'] / superstore['Sales'] * 100).round(2)
print("\n1️⃣ New 'Margin%' column created:")
print(superstore[['Sales', 'Profit', 'Margin%']].head(10))
print(f"\nMargin% summary:")
print(superstore['Margin%'].describe().round(2))

# Q2: Which region has the highest average profit?
print("\n2️⃣  Average Profit by Region:")
region_profit = superstore.groupby('Region')['Profit'].mean()
print(region_profit.round(2).sort_values(ascending=False))
best_region = region_profit.idxmax()
best_region_profit = region_profit.max()
print(f"\n🏆 Best region: {best_region} (avg profit: ${best_region_profit:.2f})")

# Q3: Filter for profitable transactions
print("\n3️⃣ Profitable Transactions (Profit > 0):")
profitable = superstore[superstore['Profit'] > 0]
print(f"Total transactions: {len(superstore)}")
print(f"Profitable transactions: {len(profitable)}")
print(f"Profitability rate: {len(profitable) / len(superstore) * 100:.1f}%")

print("\nProfitable transactions by region:")
print(profitable['Region'].value_counts())

print("\n⚠️ Unprofitable transactions (Profit ≤ 0):")
unprofitable = superstore[superstore['Profit'] <= 0]
print(f"Count: {len(unprofitable)}")
print("Top 5 losses:")
print(unprofitable.nsmallest(5, 'Profit')[['Product Name', 'Sales', 'Profit', 'Region']])
```
